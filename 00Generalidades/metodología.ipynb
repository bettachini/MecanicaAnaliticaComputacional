## Nombre
__Mecánica general__ o __Mecánica analítica computacional__

El primero, el que figura en el plan actual de la Ingeniería Mecánica, no transmite absolutamente nada sobre su temática. No hay nada más ``general``, es decir inespecífico, que adjuntar el adjetivo ``general`` a un nombre. 

En tanto que en el segundo cada palabra tiene su razón:
- __Mecánica__: rama de la física que estudia el movimiento de los cuerpos y su evolución en el tiempo.
  - Con esta herramienta creará una simplificación que modelara el aspecto de un dipositivo que sea de interés. 
- __Analítica__: rama de la matemática que estudia los métodos de resolución de problemas mediante el análisis matemático.
  - Con esta herramienta, y las del álgebra, obtendrá ecuaciones diferenciales que permitan predecir factores que afectan al sistema modelado.
- __Computacional__: puesta en práctica de métodos de resolución de problemas mediante el uso de computadoras.
  - La computadora resolverá los cálculos de forma que ya no querrá volver a realizarles a mano. Asimismo generará representaciones gráficas de resoluciones numéricas. Algo mucho más ilustrativo de lo que hace o sucede con el dispositivo mecánico que unas meras ecuaciones diferenciales.

### Encuentro sincrónico  
Se destina a resuelver los ejercicios de la guía.
La idea es contar con asesoramiento del docente al momento de requerir una consulta.
Asimismo puede evacuar inquietudes sobre la teoría que surgan durante la resolución de los ejercicios.

__No habrá__ _lección_ o _clase magistral_ por parte del docente.
Una mera lectura de lo que se escribió en un pizarrón o en una presentación proyectada __no es un buen uso del poco tiempo__ en que alumnos y docentes pueden conversar.

A las 19:30 hs de cada jornada de encuentro sincrónico cierra la entrega de resoluciones de la guía de ejercicios de la temática anterior.


### Antes  
La teoría se lee completa __antes__ del encuentro sincrónico.
Por supuesto que surgen dudas al buscar entender algo nuevo.
Estas tendrán respuesta durante la semana (¡__no__ hay que esperar al encuentro sincrónico!) tras expresarles en el equipo de __Teams__ de la asignatura.
Este tiene tantos canales como temáticas figuran en el cronograma.

La mejor manera de saber si se entendió algo es ponerlo en práctica.
Es ideal comenzar la resolución de la guía de ejercicios __antes__ del encuentro sincrónico.
Todo inconveniente que surga en hacer esto también tendrá respuesta vía __Teams__. 


### Luego  
Es probable que algún ejercicio quede sin finalizar durante el encuentro sincrónico.
Como deben entregarse todos y cada uno de los ejercicios de la guía, estos deben terminarse luego.
También es posible que surjan dudas nuevas dudas sobre la teoría.
Nuevamente toda consulta puede realizarse 24/7 en la plataforma __Teams__.
