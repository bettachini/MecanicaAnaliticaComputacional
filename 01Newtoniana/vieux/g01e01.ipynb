## Ejercicio 1
Este ejercicio es el ejemplo 2.10 de
Stephen T. Thornton, y Jerry B. Marion. Classical Dynamics of Particles and Systems. Fifth. Thomson Brooks/Cole, 2004.

### Enunciado
![Marion Fig 2.12](marionFig2_12.png)

La fuerza de Lorentz es la ejercida a una particular de carga eléctrica $q$ por un campo eléctrico $\vec{E}$ y magnético $\vec{B}$:
$$
        \vec{F}= q \left( \vec{E}+ \vec{v} \times \vec{B} \right).
$$
La figura muestra esquemáticamente una trayectoria y las condiciones $\vec{v}_0, \vec{B}=\mathrm{cte}.$ que le dieron lugar.
Halle las ecuaciones de tal dinámica.

### Resolución

#### Todo nos lleva a Newton...
Se comienza planteando la 2.a ley de la dinámica de Newton
$$
\vec{F}= q \left( \vec{E}+ \vec{v} \times \vec{B} \right) = m \vec{a}.
$$

#### Desarrollando producto vectorial
Se calcula el término que tiene el producto vectorial $\vec{v} \times \vec{B}$, y para eso se necesita una expresión de $\vec{v}$.

Si la posición es
$$
\vec{r} = x \hat{x} + y \hat{y} + z \hat{z},
$$
entonces
$$
\dot{\vec{r}} =  \frac{\mathrm{d}}{\mathrm{d} t} \vec{r}  = \dot{x} \hat{x} + \dot{y} \hat{y} + \dot{z} \hat{z}.
$$

En el dibujo puede verse que $\vec{B}$ solo tiene componente en $\hat{y}$, es decir $\vec{B} = B_0 \hat{y}$, donde $B_0 =\mathrm{cte}$.
Por tanto el producto vectorial buscado es

$$
\vec{v} \times \vec{B} = 
\begin{vmatrix}
\hat{x} & \hat{y} & \hat{z} \\
\dot{x} & \dot{y} & \dot{z} \\
B_x & B_y & B_z \\
\end{vmatrix}
=
\begin{vmatrix}
\hat{x} & \hat{y} & \hat{z} \\
\dot{x} & \dot{y} & \dot{z} \\
0 & B_0 & 0 \\
\end{vmatrix}
= - \dot{z} B_0 \hat{x} + \dot{x} B_0 \hat{z}
$$

#### Ahora si, fuerza es masa por aceleración, y entonces...
Aún hay que obtener la aceleración
$$
\ddot{\vec{r}} =  \frac{\mathrm{d}^2}{\mathrm{d} t^2} \vec{r} = \frac{\mathrm{d}}{\mathrm{d} t} \vec{v}  = \ddot{x} \hat{x} + \ddot{y} \hat{y} + \ddot{z} \hat{z}.
$$

Como aquí no hay campo eléctrico, es decir $\vec{E} = 0$, la 2.a ley queda
$$
\vec{F}
= - q \dot{z} B_0 \hat{x} + q \dot{x} B_0 \hat{z}
= m \ddot{x} \hat{x} + m \ddot{y} \hat{y} + m \ddot{z} \hat{z}
= m \ddot{\vec{r}}
$$

#### Resolveremos en cada eje. Empezamos por $\hat{y}$.
Allí $m \ddot{y} = 0 \implies \dot{y}= \mathrm{cte} = \dot{y}_0$.
Para obtener la dinámica en esta variables, es decir $y= y(t)$ debe resolverse la ecuación diferencial

$$
\begin{aligned}
\dot{y} = \frac{\mathrm{d}}{\mathrm{d} t} y &= \dot{y}_0\\
\int \mathrm{d}y &= \dot{y}_0 \int \mathrm{d} t\\
y - y_0 &= \dot{y}_0 \left(t - t_0\right),
\end{aligned}
$$

donde \(y_0 = y(t_0\) es una condición inicial que debe conocerse para determinar una trayectoria de todas las posibles.

#### Los ejes $\hat{x}, \hat{z}$ están entrelazados
$$
\left\{
\begin{array}{l}
\ddot{x} = -\frac{q}{m} B_0 \dot{z}\\
\ddot{z} = +\frac{q}{m} B_0 \dot{x}\\
\end{array}
\right.
$$

Podemos derivar una de ellas para luego sustituir una derivada en la otra.
Por ejemplo si derivo la de $x$ obtengo

$$
\begin{aligned}
\dddot{x} 
= \frac{\mathrm{d}^3}{\mathrm t^3} x
= -\frac{q}{m} B_0 \ddot{z} 
= -\frac{q}{m} B_0 \left( \frac{q}{m} B_0 \dot{x} \right)
& = - \left( \frac{q}{m} B_0 \right)^2 \dot{x}\\
\dddot{x} + \left( \frac{q}{m} B_0 \right)^2 \dot{x} &= 0
\end{aligned}
$$

Debe resolverse esta ecuación diferencial:
- lineal: polinomio de $x$ o u sus derivadas,
- homogénea: todos sus términos dependen de $x$ o sus derivadas.

Proponiendo una solución $x = \operatorname{e}^{\gamma t}$ se obtiene

$$
\left[ \gamma^3 + \left( \frac{q}{m} B_0 \right)^2 \gamma \right] \operatorname{e}^{\gamma t} = 0
\implies \gamma^2 = - \left( \frac{q}{m} B_0 \right)^2
\implies \gamma = \pm i \frac{q}{m} B_0.
$$

Como hoy dos posibles $\gamma$ que dan dos soluciones distintas, la solución es una combinación lineal de ambas

$$
x(t) = A \operatorname{e}^{+i \frac{q}{m} B_0 t} + B \operatorname{e}^{-i \frac{q}{m} B_0 t}.
$$

Como se habrá visto en un curso de cálculo avanzado, a través de la identidad de Euler puede escribirse tal solución como combinación lineal de armónicas

$$
x(t) = a_1 \cos{\left(\frac{q}{m} B_0 t \right) } + b_1 \sin{\left(\frac{q}{m} B_0 t \right) }.
$$

Evidentemente podría haberse hecho lo mismo para $z$ por lo que se hubiera obtenido

$$
z(t) = a_2 \cos{\left(\frac{q}{m} B_0 t \right) } + b_2 \sin{\left(\frac{q}{m} B_0 t \right) },
$$
con coeficientes $a_2,b_2$ en principio independientes de los de $x(t)$.

#### Sé estableció que $x$ y $z$ están entrelazadas
Recordemos que

$$
\left\{
\begin{array}{l}
\ddot{x} = -\frac{q B_0}{m} \dot{z}\\
\ddot{z} = +\frac{q B_0}{m} \dot{x}\\
\end{array}
\right.
$$
Esto es útil para encontrar relaciones entre los coeficientes $a_1, b_1, a_2, b_2$.
    
Si se derivan dos veces en el tiempo las soluciones se obtienen las mismas multiplicadas por $- \left( \frac{q B_0}{m} \right)^2$, e.g.

$$
    \ddot{x} = - \left( \frac{q B_0}{m} \right)^2 x,
$$
se utiliza esto en la primer expresión anterior y se reemplaza $x$ y $\dot{z}$ con las expresiones de sus soluciones armónicas

$$
\begin{aligned}
\ddot{x}(t) = - \left( \frac{q B_0}{m} \right)^2 \left[ x(t) \right] 
&= -\frac{q B_0}{m} \left[ \dot{z}(t) \right]\\
- \left( \frac{q B_0}{m} \right)^2 \left[ a_1 \cos{\left(\frac{q B_0}{m} t \right) } + b_1 \sin{\left(\frac{q B_0}{m} t \right) } \right] 
&= -\left( \frac{q B_0}{m} \right)
\left[
- \left( \frac{q B_0}{m} \right) a_2 \sin{\left(\frac{q B_0}{m} t \right) } + b_2 \left( \frac{q B_0}{m} \right) \cos{\left(\frac{q B_0}{m} t \right) }
\right]
\\
\left[ a_1 \cos{\left(\frac{q B_0}{m} t \right) } + b_1 \sin{\left(\frac{q B_0}{m} t \right) } \right] 
&= 
\left[
- a_2 \sin{\left(\frac{q B_0}{m} t \right) } + b_2 \cos{\left(\frac{q B_0}{m} t \right) }
\right].
\end{aligned}
$$
Como esta igualdad ha de cumplirse $\forall t$, pueden usarse distintos tiempos para encontrar las relaciones entre los coeficientes.

Así $t=0 \implies \cos{(0)} = 1$ y $\sin{(0)} = 0 \implies a_1 = b_2$.
En contrapartida habrá algún $t$ para el cuál $\left(\frac{q B_0}{m} t \right) = \frac{\pi}{2} \implies \cos{(\frac{\pi}{2})} = 0$ y $\sin{(\frac{\pi}{2})} = 1 \implies b_1 = - a_2$.

Esto permite escribir ambas soluciones armónicas $x(t), z(t)$ en función de los mismos coeficientes

$$
\left\{
\begin{array}{l}
x(t) - x_0 = a_1 \cos{\left(\frac{q}{m} B_0 t \right) } + b_1 \sin{\left(\frac{q}{m} B_0 t \right) }\\
y(t) - y_0 = \dot{y}_0 t\\
z(t) - z_0 = -b_1 \cos{\left(\frac{q}{m} B_0 t \right) } + a_1 \sin{\left(\frac{q}{m} B_0 t \right) }\\
\end{array}
\right.
$$

#### Para finalizar
Para determinar una de todas las posibles trayectorias debe conocerse una condición inicial, tanto en posición como en velocidad.
Si por ejemplo en $t=0$ se asume que se encuentra en el origen de coordenadas y solo hay velocidad inicial en $\hat{y}$ y $\hat{z}$, se tendra que

$$
\dot{x}=0 \implies 0 = \left(\frac{q B_0}{m} \right) b_1 \implies b_1 = 0 $$
y

$$
\dot{z}= \dot{z}_0 = \left(\frac{q B_0}{m} \right) a_1 \implies 
\left\{
\begin{array}{l}
x(t) = \frac{m \dot{z}_0}{q B_0} \cos{\left(\frac{q B_0}{m} t \right) }\\
y(t) = \dot{y}_0 t\\
z(t) = \frac{m \dot{z}_0}{q B_0} \sin{\left(\frac{q}{m} B_0 t \right) }\\
\end{array}
\right.
$$
