# Metodología de evaluación de entregas de ejercicios

Para aprobar la materia DEBEN realizar las entregas semanales de ejercicios.

Cada unidad de la materia tiene una guía de ejercicios, de los cuales varios serán solicitados para la entrega, éstos les aparecerán en _Teams_ bajo _Tareas_.

Las clases son para consultar y avanzar en los ejercicios, no duden en preguntarnos todo lo que requieran, de forma tal de que puedan aprovecharnos al máximo para evacuar sus dudas.

El día que se presenta la unidad podrán comenzar a mirarlos y resolverlos, tendrán la siguiente clase para realizar consultas adicionales y deberán ser entregados indefectiblemente antes de que comience la siguiente.

Es muy importante que aprovechen las clases para las consultas y que también nos escriban por Teams durante la semana consultando lo que no les salga.




__El esquema de entregas es entonces:__

Unidad N (clase N), guía N $→$ clase N+1 pueden consultar $→$ antes del comienzo de la clase N + 2 todos los ejercicios solicitados deben ser entregados.


Obsérvese que las entregas son por _Teams_ por lo que está predefinido el momento exacto en que vence la entrega: 19:00 de la clase N + 2.


Una vez vencida la entrega se desaprueba ese ejercicio. Por ello es importante que intenten avanzar en ellos, no es necesario que estén perfectos, una vez los envíen se los corregiremos y les pediremos las modificaciones necesarias para que puedan avanzar hasta que lo terminen satisfactoriamente. Cuando termine la cursada entonces deberían haber logrado hacer TODOS los ejercicios que les enviamos como _Tarea_.

Si por no entregar a tiempo un ejercicio lo desaprobaron, de todas formas les vamos a pedir que lo realicen para la semana siguiente y entreguen. En caso de no hacerlo se sumará otro desaprobado.

La nota de las _Tareas_ será un promedio de todos las notas de los ejercicios entregados y formará parte de la nota final de la materia.
