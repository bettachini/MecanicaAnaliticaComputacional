## Compartir los archivos de las soluciones a las guías de ejercicios

Tanto para entregar las soluciones a los ejercicios semanales como para enviarnos entregas parciales con consultas vamos a necesitar que compartan con permisos apropiados los notebooks.

Para ello:
1. Ir a https://colab.research.google.com
2. Acceder

00.png

3- Seleccionar si subir un notebook que tengan localmente o seleccionarlo de su google drive.

01.png

4- Compartir el archivo (ver flecha en la siguiente figura)

03.png

Se les desplegará una ventana donde deben modificar la restricción de acceso general (ver flecha en siguiente figura).

04.png

Aquí pueden elegir agregar nuestros email en el cuadro o directamente seleccionar "Cualquier persona que tenga el vínculo".

05.png

__IMPORTANTE__ Finalmente deben modificar la forma en que lo comparten (1) a "Editor" (2) en la figura siguiente.

De esta forma podremos incluir comentarios en los notebooks que nos evíen y también modificar y adicionar código y texto.



06.png

5- Una vez realizados estos pasos deben seleccionar

07.png

Este vínculo es el que utilizarán para entregar el ejercicio en el Tams. Para ello abren su página de Teams, van a "Tareas", como se indica en la siguiente imagen:

08.png

Ya en "Tareas" deben ir al ejercicio particular que desear enviar, y allí tendrán una pantalla muy parecida a la que se muestra en la figura de abajo.

Allí seleccionan el ícono de un clip, luego "vínculo" y pegan allí el link que copiaron del Google Colab.

__IMPORTANTE__ Finalmente deben "Entregar" la tarea utilizando el botón correspondiente, hasta tanto no lo hagan no nos llegará a nosotros.

09.png
