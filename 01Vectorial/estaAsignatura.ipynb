## ¿Por qué esta asignatura se llama _Mecánica analítica computacional_?

- El nombre oficial es _Mecánica General_
    - En una carrera de _Ingeniería mecánica" no podría haber uno menos explicativo...
    - Por esa razón en el _nuevo plan_ se llama **Mecánica analítica computacional**
- Mecánica Analítica
    - La física de un sistema de cuerpos puede resolverse en forma **vectorial**, como hicieron en Física I,
    - o haciendo balance de **energías** (cinética, potencial, etc), como también hicieron en FI.
    - Este último enfoque, el **analítico** es mucho más poderoso de lo que vieron en FI.

- De libros (**papel**) el profesor **copia** teoría y ejemplos de resolución al **pizarrón**.
- Del **pizarrón** el alumno copia a su cuaderno de **papel** ambos
- Cada vez que hace un ejercicio vuelve a resolver en **papel** modelos y cuentas similares.

## Programar $\neq$ escribir código

- **Programar**: Saber cómo se puede llegar a resolver un problema, analizar y comparar alternativas y elegir la mejor considerando contexto, restricciones, etc.
- **Escribir código**: Plasmar una solución en un lenguaje de programación particular.
