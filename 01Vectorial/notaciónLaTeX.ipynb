## Una "nota" sobre la "notaci√≥n" (¬°cuac!).

Los cuadernos (notebook) **Jupyter** permiten utilizar la escritura de expresiones matem√°ticas en notaci√≥n [$\LaTeX$](https://www.latex-project.org), un sistema de preparaci√≥n de documentos en el que todo, inclusive las ecuaciones, se escribe como texto plano.

Por ejemplo el s√≠mbolo que designa la velocidad angular, la **phi** del alfabeto griego, se puede lograr escribiendo `\varphi` logrando que se muestre $\varphi$, o con su otra graf√≠a `\phi` logrando $\phi$.

$\LaTeX$ permite escribir m√°s cosas en formato matem√°tico que s√≠mbolos sueltos.
Expresiones matem√°ticas complejas, e.g. $\sqrt{ \frac{\sin(\alpha^2 ) }{2} }$ pueden presentarse en la l√≠nea de texo englobando lo escrito entre un par de `$`.
Al ejecutar la celda `markdown` se reemplaza lo escrito en notaci√≥n $\LaTeX$ por el equivalente en la notaci√≥n matem√°tica usual.

Si la expresi√≥n es larga, o central en lo que se explica, puede escribirse en una l√≠nea aparte englobandola en un par de `$$` y aparecer√° centrada:
$$
    \sqrt{ \frac{\sin(\alpha^2 ) }{2} }.
$$
