# Enunciado
Un frisbee tiene su masa $M$ distribuida en forma uniforme en su radio $R$.
Mientras gira manteniendo su horizontalidad una araña que estaba originalmente en su centro camina en dirección radial a velocidad constante $v_a$.  
1. Escribir la expresión velocidad y aceleración de la araña consideradas desde un sistema de coordenas fijas al suelo.  
1. ¿Cuál es la $\omega$ del frisbee respecto a la original cuando la araña esté a una distancia $d$ de su centro?  
1. ¿Qué torque debiera ejercerse al frisbee si se quisiera que su $\omega$ fuera constante?  
¿Qué fuerza de _coriolis_ siente la araña en tal caso?

# Un poco de teoría

## Posición de la araña
Respecto a un sistema de referencia fijado al frisbee la araña estará en 
$$
\vec{r}_{a/f}  (t)= (v_a t, 0, 0)
$$ en coordenadas cilíndricas $(\rho, \varphi, z)$.
Ojo, como los ejes de este sistema giran con el frisbee sería un error adicionarle la rotación del frisbee en la dirección $\hat{\varphi}$. Debe quedar claro que
$$
\vec{r}_{a/f} (t) \neq (v_a t, \omega t, 0).
$$

Si queremos escribir la posición respecto a un sistema fijo al suelo $\vec{r}_{a/o}$ debemos considerar primero la coordenada del frisbee respecto al origen de coordenadas de este sistema $\vec{r}_{f/o}$ y entonces
$$
\vec{r}_{a/o} = \vec{r}_{a/f} + \vec{r}_{f/o}.
$$

$$
\vec{r}_{f/o} = x \hat{x}+ y \hat{y},

La araña siempre se mueve sobre $\hat{i}$ en su movimiento radial sobre el frisbee

$$
\vec{r}_{a/f} = \rho \hat{i}, 
$$
así escribimos la posición de la araña

$$
\boxed{
\begin{aligned}
\vec{r}_{a/o} &= \vec{r}_{a/f} + \vec{r}_{f/o} \\
&= \rho \hat{i}+ x \hat{x}+ y \hat{y}.
\end{aligned}
}
$$
Pero como la dirección $\hat{i}$ varía con $t$ habría que escribirla en relación a la del sistema fijo al suelo.
En el dibujo se muestra que puede escribirse $\hat{i}$ en función del ángulo que forma con $\hat{y}$ y por tanto

$$
\hat{i} = \cos{\phi} \hat{x}+ \sin{\phi} \hat{y},
$$
con lo que finalmente se obtiene

$$
\vec{r}_{a/o} = \rho \left( \cos{\phi} \hat{x}+ \sin{\phi} \hat{y} \right) + x \hat{x}+ y \hat{y}
= \left( \rho \cos{\phi}+ x \right) \hat{x}+ \left( \rho \sin{\phi} + y \right) \hat{y}.
$$

## Velocidad de la araña

$$
\begin{aligned}
\vec{v}_{a/o} &= \frac{\mathrm{d}}{\mathrm{d}t} \vec{r}_{a/o}
= \frac{\mathrm{d}}{\mathrm{d}t} \left( \rho \hat{i}+ x \hat{x}+ y \hat{y} \right)\\
& = \frac{\mathrm{d}}{\mathrm{d}t} \left( \rho \right) \hat{i} + \rho \frac{\mathrm{d}}{\mathrm{d}t}\left( \hat{i} \right)
+ \frac{\mathrm{d}}{\mathrm{d}t} \left( x \right) \hat{x}
+ \frac{\mathrm{d}}{\mathrm{d}t} \left( y \right) \hat{y}
\end{aligned}
$$
recordemos que $\hat{x}, \hat{y}$ no varían en el $t$, pero si lo hace $\hat{i}$

$$
\frac{\mathrm{d}}{\mathrm{d}t}\left( \hat{i} \right) = 
\frac{\mathrm{d}}{\mathrm{d}t}\left( \cos{\phi} \hat{x}+ \sin{\phi} \hat{y}
 \right) =
\dot{\phi} \left(- \sin{\phi}  \hat{x} + \cos{\phi} \hat{y} \right),
$$
que como $\hat{j}= - \sin{\phi} \hat{x} + \cos{\phi} \hat{y}$ nos permite escribir que $\frac{\mathrm{d}}{\mathrm{d}t}\left( \hat{i} \right) = \dot{\phi} \hat{j}$.

Así la velocidad queda

$$
\boxed{
\vec{v}_{a/o} = \dot{\rho} \hat{i}+ \rho \dot{\phi} \hat{j} + \dot{x} \hat{x} + \dot{y} \hat{y}
}
= \left( \dot{\rho} \cos{\phi} - \rho \dot{\phi} \sin{\phi} + \dot{x} \right) \hat{x}
+ \left( \dot{\rho} \sin{\phi} + \rho \dot{\phi} \cos{\phi} + \dot{y} \right) \hat{y}
$$

## Aceleración de la araña

$$
\vec{a}_{a/o}
= \frac{\mathrm{d}}{\mathrm{d}t}\left( \vec{v}_{a/o} \right)
= \ddot{\rho} \hat{i} + \dot{\rho} \frac{\mathrm{d}}{\mathrm{d}t}\left( \hat{i} \right) 
+ \dot{\rho} \dot{\phi} \hat{j} + \rho \ddot{\phi} \hat{j}
+ \rho \dot{\phi} \frac{\mathrm{d}}{\mathrm{d}t}\left( \hat{j} \right)
+ \ddot{x} \hat {x} + \ddot{y} \hat{y},
$$
y puesto que $\frac{\mathrm{d}}{\mathrm{d}t}\left( \hat{j} \right)= - \dot{\phi} \left( \cos{\phi} \hat{x}+ \sin{\phi} \hat{y} \right)= - \dot{\phi} \hat{i}$,

$$
\begin{aligned}
\vec{a}_{a/o}
&=  \ddot{\rho} \hat{i} + \dot{\rho} \dot{\phi} \hat{j}
+ \dot{\rho} \dot{\phi} \hat{j} + \rho \ddot{\phi} \hat{j}
- \rho \dot{\phi} \dot{\phi} \hat{i}
+ \ddot{x} \hat {x} + \ddot{y} \hat{y} \\
&= \left( \ddot{\rho} - \rho \dot{\phi} \dot{\phi} \right) \hat{i}
+ \left( \dot{\rho} \dot{\phi} + \dot{\rho} \dot{\phi} + \rho \ddot{\phi} \right) \hat{j}
+ \ddot{x} \hat {x} + \ddot{y} \hat{y}.
\end{aligned}
$$
Obtenemos entonces

$$
\boxed{
\vec{a}_{a/o}=
\ddot{\rho} \hat{i}
- \rho \dot{\phi}^2 \hat{i}
+ \rho \ddot{\phi} \hat{j} 
+ 2 \dot{\rho} \dot{\phi} \hat{j}
+ \ddot{x} \hat {x} + \ddot{y} \hat{y}
}.
$$
Cada uno de los primeros cuatro términos es una aceleración relacionada a la rotacíon que recibe un nombre partícular.
Los dos primeros son los de las aceleración en cilíndricas:

1. **Lineal**: Que es la convencion unidimensional, en la dirección radial en el frisbee.  
2. **Centrípeta**: La necesaria para mantener la araǹa en el frisbee noten que el signo hace que apunte al centro del mismo.

en tanto que el tercero y cuarto surgen de estár analizando la aceleración desde un sistema inercial no centrado en el cuerpo en rotación:  

3. **Euleriana**: Cuanto mayor el alejamiento del centro $\rho$ un aumento de de la velocidad angular $\ddot{\phi}$ la araña percibirá un fuerza inercial que la empuja en el sentido opuesto a la rotación del frisbee. E.g. cuando un calesita empieza a girar un nene  en un caballito se siente tirado "hacia atrás", y si se detiene abruptamente "sale volando hacia adelante".
4. **Coriolis**: Cuanto más rápido la araña camina hacia el borde exterior percibe una la fuerza inercial adicional que la _empuja_ en el sentido opuesto a la rotación del frisbee.

Estas últimas se las denomina _ficticias_ pues son percibidas solo en relación a un sistema que acelera, y los términos que figuran aquí son las aceleraciones que tiene que tener la araña para **compensar las correspondientes fuerzas ficticias**, así que tienen el signo opuesto a lo que percibe la araña (3.a ley de Newton).

Un ejemplo clásico de **fuerza ficticia** no relacionada a una rotación es el _empujón_ hacia adelante que percibimos respecto a un colectivo si este frena abruptamente.
Para un observador en la calle los pasajeros siguieron con la velocidad del colectivo, sin sufrir aceleración (fuerza), fué solo la masa del colectivo la que sufrió una fuerza que lo detuvo.

## Una forma más rápida de calcular esto: el teorema del transporte

$$
\boxed{
\vec{v}_O = \frac{\mathrm{d}}{\mathrm{d} t} \vec{r}_O = \frac{\mathrm{d}}{\mathrm{d} t} \vec{r}_P + \vec{\Omega}_{P/O} \times \vec{r}_P = \vec{v}_P + \vec{\Omega}_{P/O} \times \vec{r}_P
},
$$
siendo $P$ el sistema fijado al sistema no inercial que rota con $\vec{\Omega}_{P/O}$y $O$ el inercial (e.g. sujeto al suelo), por lo que $\vec{r}_P$ es la posición del centro de este sistema desde el sistema $O$, y $\vec{r}_O$ la del punto cuya velocidad en este mismo sistema se desea determinar.

>**Referencia: teorema del trasporte**  
$\S 31$ Velocidad angular  
Lev Davidovich Landau, y E. M. Lifshitz. Mecánica. 2.a. Curso de física teórica. Reverté, 1994.

### Velocidad con el teorema del transporte
Se parte de la posición de la araña:

$$
\vec{r}_{a/o} = \vec{r}_{a/f} + \vec{r}_{f/o} = \rho \hat{i}+ x \hat{x}+ y \hat{y}.
$$
Para hallar la **velocidad en el sistema fijo al suelo**

$$
\vec{v}_{a/o}
= \frac{\mathrm{d}}{\mathrm{d}t} \left( \vec{r}_{a/o} \right)_o
= \frac{\mathrm{d}}{\mathrm{d}t} \left( \vec{r}_{a/f} \right)_o  + \frac{\mathrm{d}}{\mathrm{d}t} \left( \vec{r}_{f/o} \right)_o
= \frac{\mathrm{d}}{\mathrm{d}t} \left( \rho \hat{i} \right)_o + \left( x \hat{x}+ y \hat{y} \right),
$$
pero ahora no vamos a desarmar $\hat{i}$, sino que vamos a aplicar el teorema del transporte

$$
\frac{\mathrm{d}}{\mathrm{d}t} \left( \rho \hat{i}\right)_o = \frac{\mathrm{d}}{\mathrm{d}t} \left( \rho \hat{i}\right)_f + \left( \vec{\omega}_{f/o} \times \rho \hat{i} \right)_f ,
$$
que nos evita preocuparnos de preocuparnos en $\frac{\mathrm{d}}{\mathrm{d}t} \left( \hat{i}\right)_f$ pues $\hat{i}$ no cambia de dirección en el sistema fijo al frisbee.

Aquí la vida se nos hace sencilla porque $\vec{\omega}_{f/o}= \dot{\phi} \hat{z} =  \dot{\phi} \hat{k}$ siendo $\hat{k}$ el que completa la triada del sistema fijo al frisbee $\hat{i}, \hat{j}$, y que es siempre paralelo a $\hat{z}$.
Así

$$
\frac{\mathrm{d}}{\mathrm{d}t} \left( \rho \hat{i}\right)_o
= \frac{\mathrm{d}}{\mathrm{d}t} \left( \rho \hat{i}\right)_f + \left( \vec{\omega}_{f/o} \times \rho \hat{i} \right)_f
= \dot{\rho} \hat{i}+ \dot{\phi} \hat{k} \times \rho \hat{i}
= \dot{\rho} \hat{i}+ \rho \dot{\phi} \hat{j},
$$
y obtenemos el mismo resultado que antes para la velocidad

$$
\begin{aligned}
\vec{v}_{a/o} &= \frac{\mathrm{d}}{\mathrm{d}t} \left( \vec{r}_{a/o} \right)_o \\
&= \frac{\mathrm{d}}{\mathrm{d}t} \left( \vec{r}_{a/f} \right)_o  + \frac{\mathrm{d}}{\mathrm{d}t} \left( \vec{r}_{f/o} \right)_o \\
&= \frac{\mathrm{d}}{\mathrm{d}t} \left( \rho \hat{i}\right)_f + \left( \vec{\omega}_{f/o} \times \rho \hat{i} \right)_f + \frac{\mathrm{d}}{\mathrm{d}t} \left( \vec{r}_{f/o} \right)_o \\
&= \dot{\rho} \hat{i}+ \rho \dot{\phi} \hat{j} + \dot{x} \hat{x} + \dot{y} \hat{y} \\
\end{aligned}
$$

## Aceleración con el teorema del transporte
Para completar hagamos lo mismo para llegar a la **aceleración medida en el sistema fijo al suelo**.
Recordamos que

$$
\begin{aligned}
\vec{a}_{a/o} 
&= \frac{\mathrm{d}}{\mathrm{d}t} \left( \vec{v}_{a/o} \right) =
\frac{\mathrm{d}}{\mathrm{d}t} \left( \vec{v}_{a/f} \right)_o
+ \frac{\mathrm{d}}{\mathrm{d}t} \left( \vec{v}_{f/o} \right)_o \\
&=
\frac{\mathrm{d}}{\mathrm{d}t} \left(  \vec{v}_{a/f}+ \vec{\omega}_{f/o} \times  \vec{r}_{a/f} \right)_o 
+ \frac{\mathrm{d}}{\mathrm{d}t} \left( \vec{v}_{f/o} \right)_o \\
&= \frac{\mathrm{d}}{\mathrm{d}t} \left(
\vec{v}_{a/f}+ \vec{\omega}_{f/o} \times \vec{r}_{a/f}
\right)_f
+ \left( \vec{\omega}_{f/o} \times \left(
\vec{v}_{a/f}+ \vec{\omega}_{f/o} \times \vec{r}_{a/f} 
\right) \right)_f 
+ \frac{\mathrm{d}}{\mathrm{d}t} \left( \vec{v}_{f/o} \right)_o \\
&= \vec{a}_{a/f}
+ \left( \dot{\vec{\omega}}_{f/o} \times \vec{r}_{a/f} \right)
+ \left( \vec{\omega}_{f/o} \times \vec{v}_{a/f} + \vec{\omega}_{f/o} \times \vec{v}_{a/f} \right)
+ \left( \vec{\omega}_{f/o} \times \left( \vec{\omega}_{f/o} \times \vec{r}_{a/f} \right) \right)
+ \vec{a}_{f/o}
\end{aligned}
$$
donde el 2.do término es la aceleración euleriana, el 3.ero el de _coriolis_ y el 4.to la aceleracíon centrípeta.
Desarrollamos los cinco

$$
\begin{aligned}
\vec{a}_{a/o}
&= \left( \ddot{\rho} \hat{i}\right)
+ \left( \ddot{\phi} \hat{k} \times \rho \hat{i} \right)
+ \left(2 \dot{\phi} \hat{k} \times \dot{\rho} \hat{i} \right)
+ \left( \dot{\phi} \hat{k} \times \left( \dot{\phi} \hat{k} \times \rho \hat{i} \right) \right)
+  \left( \ddot{x} \hat{x} +  \ddot{y} \hat{y} \right) \\
&= \left( \ddot{\rho} \hat{i}\right)
+ \left( \rho \ddot{\phi} \hat{j} \right)
+ \left(2 \dot{\rho} \dot{\phi} \hat{j} \right)
+ \left( \dot{\phi} \hat{k} \times \left( \rho \dot{\phi} \hat{j} \right) \right) 
+ \left( \ddot{x} \hat{x} +  \ddot{y} \hat{y} \right) \\
&= \left( \ddot{\rho} \hat{i}\right)
+ \left( \rho \ddot{\phi} \hat{j} \right)
+ \left(2 \dot{\rho} \dot{\phi} \hat{j} \right)
+ \left(- \rho \dot{\phi}^2 \hat{i} \right) 
+ \left( \ddot{x} \hat{x} +  \ddot{y} \hat{y} \right) \\
\end{aligned}
$$

Para resumir como se obtiene la aceleración vista desde un sistema inercial de una partícula en un sistema no inercial ayudados por el teorema del transporte

$$
\begin{aligned}
\vec{a}_O
& = \frac{\mathrm{d}}{\mathrm{d} t} \vec{v}_O
= \frac{\mathrm{d}}{\mathrm{d} t} \left(
\vec{v}_P + \vec{\Omega}_{P/O} \times \vec{r}_P
\right)
+ \vec{\Omega}_{P/O} \times \left(
\vec{v}_P + \vec{\Omega}_{P/O} \times \vec{r}_P
\right) \\
& = \left( \vec{a}_P + \dot{\vec{\Omega}}_{P/O} \times \vec{r}_P + \vec{\Omega}_{P/O} \times \vec{v}_P \right)
  + \left( \vec{\Omega}_{P/O} \times \vec{v}_P + \vec{\Omega}_{P/O} \times \vec{\Omega}_{P/O} \times \vec{r}_P \right),
\end{aligned}
$$
que queda la larga expresión

$$
\boxed{
\vec{a}_O = \vec{a}_P + \dot{\vec{\Omega}}_{P/O} \times \vec{r}_P + 2 \vec{\Omega}_{P/O} \times \vec{v}_P + \vec{\Omega}_{P/O} \times \vec{\Omega}_{P/O} \times \vec{r}_P
}
$$
siendo $P$ el sistema fijado al sistema no inercial que rota con $\vec{\Omega}_{P/O}$ y $O$ el inercial (e.g. sujeto al suelo), por lo que $\vec{r}_P$ es la posición del centro de este sistema desde el sistema $O$, y $\vec{r}_O$ la del punto cuya velocidad en este mismo sistema se desea determinar.

# Resolución
Como haríamos este problema aprovechando lo aprendido de teoría.

## 1. Velocidad y aceleración desde un sistema fijo al suelo
**Posición de la araña** respecto al sistema sujeto al frisbee 
$$
    \vec{r}_{a/f}= \rho \hat{i}
$$ 
y respecto al sistema en el suelo 
$$
    \vec{r}_{a/o}= \vec{r}_{a/f} + \vec{r}_{f/o} = \rho \hat{i} + \left( x \hat{x}+ y \hat{y} \right).
$$

**Velocidad de la araña**

$$
\begin{aligned}
\vec{v}_{a/o} &=  \left( \frac{\mathrm{d}}{\mathrm{d} t} \vec{r}_{a/o} \right)_o
= \left( \frac{\mathrm{d}}{\mathrm{d} t}  \vec{r}_{a/f} \right)_o + \frac{\mathrm{d}}{\mathrm{d}t} \vec{r}_{f/o} \\
&= \left( \frac{\mathrm{d}}{\mathrm{d} t} \vec{r}_{a/f} \right)_f + \vec{\omega}_f \times \vec{r}_{a/f} + \frac{\mathrm{d}}{\mathrm{d}t} \vec{r}_{f/o} \\
&= \dot{\rho} \hat{i} + \dot{\phi} \hat{k} \times \rho \hat{i} + \left( \dot{x} \hat{x}+ \dot{y} \hat{y} \right) \\
&= \dot{\rho} \hat{i} + \rho \dot{\phi} \hat{j} + \left( \dot{x} \hat{x}+ \dot{y} \hat{y} \right).
\end{aligned}
$$

**Aceleración de la araña**

$$
\begin{aligned}
\vec{a}_{a/o} &=  \left( \frac{\mathrm{d}}{\mathrm{d} t} \vec{v}_{a/o} \right)_o \\
&= 
\frac{\mathrm{d}}{\mathrm{d}t} \left(  \vec{v}_{a/f}+ \vec{\omega}_{f/o} \times  \vec{r}_{a/f} \right)_o
+ \frac{\mathrm{d}}{\mathrm{d}t} \vec{v}_{f/o} \\
& =
\frac{\mathrm{d}}{\mathrm{d}t} \left( 
\vec{v}_{a/f}+ \vec{\omega}_{f/o} \times \vec{r}_{a/f}
\right)_f
+ \left( \vec{\omega}_{f/o} \times \left( 
\vec{v}_{a/f}+ \vec{\omega}_{f/o} \times \vec{r}_{a/f}
\right) \right)_f
+ \frac{\mathrm{d}^2}{\mathrm{d} t^2} \vec{r}_{f/o} \\
& =
\frac{\mathrm{d}}{\mathrm{d}t} \left( 
\dot{\rho} \hat{i} + \dot{\phi} \hat{k} \times \rho \hat{i}
\right)_f
+ \left( \dot{\phi} \hat{k} \times \left( 
\dot{\rho} \hat{i} + \dot{\phi} \hat{k} \times \rho \hat{i}
\right) \right)_f
+  \left( \ddot{x} \hat{x}+ \ddot{y} \hat{y} \right) \\
& =
\frac{\mathrm{d}}{\mathrm{d}t} \left( 
\dot{\rho} \hat{i} + \rho \dot{\phi} \hat{j}
\right)_f
+ \left( \dot{\phi} \hat{k} \times \left( 
\dot{\rho} \hat{i} + \rho \dot{\phi} \hat{j}
\right) \right)_f
+  \left( \ddot{x} \hat{x}+ \ddot{y} \hat{y} \right) \\
& =
\left( 
\ddot{\rho} \hat{i} + \dot{\rho} \dot{\phi} \hat{j}+  \rho \ddot{\phi} \hat{j}
\right)_f
+ \left( \dot{\rho} \dot{\phi} \hat{j} - \rho \dot{\phi}^2 \hat{i} \right)_f
+  \left( \ddot{x} \hat{x}+ \ddot{y} \hat{y} \right) \\
& =
  \left( \ddot{\rho} \hat{i} \right) 
+ \left( 2\dot{\rho} \dot{\phi} \hat{j} \right)
+ \left( \rho \ddot{\phi} \hat{j} \right)
+ \left( - \rho \dot{\phi}^2 \hat{i} \right)
+  \left( \ddot{x} \hat{x}+ \ddot{y} \hat{y} \right) \\
\end{aligned}
$$

## 2. $\omega_\mathrm{frisbee}$ cuando la araña esté a $d$ de su centro
- No hay fuerza externa $\implies \vec{\tau}=0 \implies \frac{\mathrm{d}}{\mathrm{d}t} \vec{L}= 0$.
El momento angular $\vec{L}$ es constante.

- El frisbee vuela horizontal $\implies \vec{\omega} = \omega_z \hat{z} \forall t \implies \vec{L}= I_{zz} \omega_z \hat{z}$.

- Si la masa de la araña $m_a$ es despreciable respecto a la del frisbee $M$ $\implies$ el centro de masa coincide con el centro del frisbe $\forall t$.
Se calculan desde ahí los momento de inercia del sistema $I_{zz}= I_\mathrm{frisbee} + I_\mathrm{araña}$.
    - Si la araña es una masa puntual $\implies I_\mathrm{araña}= m_a \rho^2$.
    - Al frisbee de radio $R$, le corresponde la de un cilíndro $I_\mathrm{frisbee}= \frac{M}{2} R^2$.

- El $\vec{L}$ cuando la araña esté a una distancia $d$ del centro, $\vec{r}_{a/f}= d \hat{i}$,  es igual al inicial
$$
\begin{aligned}
    \vec{L}(t') &= \vec{L}(t=0)\\
    \left( \frac{M}{2} R^2+ m_a d^2 \right) \omega (t') &= \frac{M}{2} R^2 \omega (t=0) .
\end{aligned}
$$

Como la araña se desplaza con velocidad radial $\dot{\vec{r}}_{a/f}= v_a \hat{i}$ en el frisbee, está en $\vec{r}_{a/f}= d \hat{i}$ en el $t'= \frac{d}{v_a }$.
En ese tiempo la velocidad angular es
$$
\boxed{
\omega \left( t=\frac{d}{v_a } \right) = \frac{\frac{M}{2} R^2}{\frac{M}{2} R^2+ m_a d^2} \omega (t=0)= \frac{1}{1 + \frac{2 m_a}{M} \left( \frac{d}{R} \right)^2} \omega (t=0)
}
$$

## 3.
### $\vec{\tau}$ si $\omega_\mathrm{frisbee} = \mathrm{constante}$

Si $\vec{\omega}$ fuera constante el incremento del momento angular sería el correspondiente al desplazamiento de la araña $\Delta \vec{L} = m_a d^2 \omega \hat{z}$.
Recordemos que $\vec{\tau} = \frac{\mathrm{d}}{\mathrm{d} t} \vec{L}$, y que si este torque aplicado fue constante en el tiempo $\vec{\tau} = \frac{\Delta \vec{L}}{\Delta t}$.
El tiempo transcurrido es $\Delta t = \frac{d}{v_a}$, así que el torque requerido para mantener una $\vec{\omega}$ constante es
$$
\boxed{
\vec{\tau}= \frac{m_a d^2 \omega}{\frac{d}{v_a}} \hat{z} = m_a d v_a \omega \hat{z}
}
$$

### Fuerza de Coriolis sobre araña
En la aceleración el término de _Coriolis_ es $\vec{a}_\mathrm{coriolis}= 2 \dot{\rho} \dot{\phi} \hat{j}$.
En este caso en que se manteniene constante la velocidad angular del frisbee constante basta con multiplicar por la masa de la araña
$$
\boxed{
\vec{F}_\mathrm{coriolis}= m_a (2 v_a \omega \hat{j} ) = 2 m_a v_a \omega \hat{j}
}.
$$
