## Momento y fuerza generalizada
_Adaptado de la sección 7.1 John Robert Taylor. Classical Mechanics. University Science Books, 2005._

Escribo el Lagrangiano en coordenadas cartesianas para una partícula en un espacio bidimensional

$$
\mathcal{L} (x,y,\dot{x}, \dot{y})= T - V = \frac{m}{2} (\dot{x}^2 + \dot{y}^2) - V(x,y).
$$

La derivada respecto a la velocidad en $\hat{x}$

$$
\frac{\partial}{\partial \dot{x}} \mathcal{L} = \frac{\partial}{\partial \dot{x}} T = m \dot{x} = p_x,
$$
es el **momento** en $\hat{x}$.

La derivada respecto de la coordenada en si 
$$
\frac{\partial}{\partial x} \mathcal{L} = - \frac{\partial}{\partial x} V = F_x,
$$
es la **fuerza** causada por el potencial $V$.

Se puede así recuperar la 2.a ley de Newton
$$
\left\{
  \begin{array}{l}
    \frac{\partial}{\partial x} \mathcal{L} =  \frac{\mathrm{d}}{\mathrm{d} t} \frac{\partial}{\partial \dot{x}} \mathcal{L}\\
    \frac{\partial}{\partial y} \mathcal{L} =  \frac{\mathrm{d}}{\mathrm{d} t} \frac{\partial}{\partial \dot{y}} \mathcal{L}
  \end{array}
\right.
\iff
\left\{
  \begin{array}{l}
    F_x = \frac{\mathrm{d}}{\mathrm{d} t} p_x\\
    F_y = \frac{\mathrm{d}}{\mathrm{d} t} p_y
  \end{array},
\right.
$$
donde los $\frac{\mathrm{d}}{\mathrm{d} t} p_i$ equivalen a $m \ddot{q}_i$, si $q_i$ son coordenadas cartesianas como $x,y$.


De usarse coordenadas generalizadas para escribir $\mathcal{L}(q_i,\dot{q}_i,t)$ estas derivadas no necesariamente son módulos de los momentos o fuerzas en el sentido vectorial, pero tienen un rol similar y por eso se llaman
-  $\frac{\partial}{\partial \dot{q}_i} \mathcal{L}$: componente $i$ de un **momento generalizado**
-  $\frac{\partial}{\partial q_i} \mathcal{L}$: componente $i$ de una **fuerza generalizada**

### Partícula en sistema cilíndrico
Veamos en este sistema de coordenadas que son estos momentos y fuerzas generalizadas.

En este sistema
$$
\mathcal{L}(\rho, \dot{\rho},\varphi, \dot{\varphi}, z, \dot{z}) = \frac{m}{2} \left( \dot{\rho}^2+ \rho^2 \dot{\varphi}^2 + \dot{z}^2 \right) - V(\rho, \varphi, z).
$$
Aquí el momento para la cordenada generalizada $\phi$
$$
\frac{\partial}{\partial \dot{\varphi}} \mathcal{L} = m \rho^2 \dot{\varphi},
$$
resulta ser el **momento angular**.

Y como de Euler-lagrange tenemos que
$$
\frac{\mathrm{d}}{\mathrm{d} t} \frac{\partial}{\partial \dot{\varphi}} \mathcal{L} = \frac{\partial}{\partial \varphi} \mathcal{L},
$$
este última _fuerza generalizada_ para la coordenada $\varphi$ resulta ser la variación en el tiempo del momento angular, es decir el **torque**,
$$
\tau = \frac{\mathrm{d}}{\mathrm{d} t} \left( m \rho^2 \dot{\varphi} \right) = - \frac{\partial}{\partial \varphi} V = \frac{\partial}{\partial \varphi} \mathcal{L}.
$$

## Fuerzas de vínculo y multiplicadores de Lagrange
_Adaptado de sección 7.10
John Robert Taylor. Classical Mechanics. University Science Books, 2005._


Una ecuación que explicite un vínculo podemos escribirla como

$$
f(x,y)= \text{const.}
$$
Por ejemplo en un péndulo simple de largo $l$ las coordenadas cartesianas están relacionadas por

$$
f(x,y)= \sqrt{x^2+ y^2}= l.
$$

En la
_sección $\S6$ de "Un destilado..."_ dedicada a la "Ecuación de Euler-Lagrange" se explica que:

La acción

$$
S = \delta \int_{t_1}^{t_2} L(q_i, \dot{q}_i, t) \delta q_i \mathrm{d}t, 
$$
debe ser estacionaria, es decir $\delta S = 0$,
que para este problema en (x,y) en que $x(t) \rightarrow x(t) + \delta x(t)$ e $y(t) \rightarrow y(t) + \delta x(t)$,
resulta en

$$
\delta S
=
 \int_{t_1}^{t_2}
    \left( 
    \frac{\partial}{\partial x} \mathcal{L} \delta x
    - \frac{\partial}{\partial \dot{x}} \mathcal{L} \delta \dot{x}
    +
    \frac{\partial}{\partial y} \mathcal{L} \delta y
    - \frac{\partial}{\partial \dot{y}} \mathcal{L} \delta \dot{y}
    \right)
 \mathrm{d}t
$$
Ya probamos que

$$
\delta S = \delta \int_{t_1}^{t_2} \left( 
    \frac{\partial}{\partial q_i} \mathcal{L}
    - \frac{\mathrm{d}}{\mathrm{d} t} \frac{\partial}{\partial \dot{q}_i} \mathcal{L}
\right) \delta q_i \mathrm{d}t, 
$$
por tanto que para un sistema descripto por las coordenadas $(x,y)$ la acción es estacionaria si

$$
\delta S
=
 \int_{t_1}^{t_2}
    \left( 
    \frac{\partial}{\partial x} \mathcal{L}
    - \frac{\mathrm{d}}{\mathrm{d} t} \frac{\partial}{\partial \dot{x}} \mathcal{L}
    \right)
    \delta x \mathrm{d}t
    + 
    \left( 
    \frac{\partial}{\partial y} \mathcal{L}
    - \frac{\mathrm{d}}{\mathrm{d} t} \frac{\partial}{\partial \dot{y}} \mathcal{L}
    \right)
\delta y \mathrm{d}t = 0
$$
Si esta expresión fuera cierta para variaciones arbitrarias de $\delta x$ y $\delta y$ resolveríamos ambas ecuaciones cada cual por su lado.
Pero aquí hay un vínculo entre ambas, que tambien podemos expresar en función de estas $\delta x$ y $\delta y$ como

$$
\delta f = \frac{\partial}{\partial x} f \delta x +  \frac{\partial}{\partial y} f \delta y = 0,
$$
que es nula pues estaba $f$ estaba igualada a una constante.

Así que sumar esta última al integrando de la última integral, aún si la multiplicaramos por una constante dependiente del tiempo $\lambda(t)$, no tiene porque cambiar el valor nulo de la integral

$$
\begin{aligned}
\delta S
=&
 \int_{t_1}^{t_2}
    \left( 
    \frac{\partial}{\partial x} \mathcal{L}
    - \frac{\mathrm{d}}{\mathrm{d} t} \frac{\partial}{\partial \dot{x}} \mathcal{L}
    + \lambda(t) \frac{\partial}{\partial x} f
    \right)
    \delta x \mathrm{d}t \\
&+ \int_{t_1}^{t_2}
    \left( 
    \frac{\partial}{\partial y} \mathcal{L}
    - \frac{\mathrm{d}}{\mathrm{d} t} \frac{\partial}{\partial \dot{y}} \mathcal{L}
    + \lambda(t) \frac{\partial}{\partial y} f
    \right)
\delta y \mathrm{d}t = 0.
\end{aligned}
$$

Y aquí viene el golpe de astucia, podemos variar $\lambda(t)$ hasta que la cantidad entre paréntesis que acompaña $\delta x$ se anule.
Como la suma de ambas integrales es igual a cero, hacer esto asegura al mismo tiempo que lo mismo sucede con la cantidad que acompaña a $\delta y$.
Es decir que tenemos ahora

$$
\left\{
  \begin{array}{l}
    \frac{\partial}{\partial x} \mathcal{L}
    - \frac{\mathrm{d}}{\mathrm{d} t} \frac{\partial}{\partial \dot{x}} \mathcal{L}
    + \lambda(t) \frac{\partial}{\partial x} f = 0\\
    \frac{\partial}{\partial y} \mathcal{L}
    - \frac{\mathrm{d}}{\mathrm{d} t} \frac{\partial}{\partial \dot{y}} \mathcal{L}
    + \lambda(t) \frac{\partial}{\partial y} f = 0,
  \end{array}
\right.
$$


Tenemos ahora solo dos ecuaciones de Euler-Lagrange  modificadas para obtener tres funciones $x(t), y(t), \lambda(t)$.
Pero como disponemos de la ecuación de la condición $f(x,y)$, tenemos tres ecuaciones y tres incógnitas:

$$
\left\{
  \begin{array}{l}
    \frac{\mathrm{d}}{\mathrm{d} t} \frac{\partial}{\partial \dot{x}} \mathcal{L} 
    = \frac{\partial}{\partial x} \mathcal{L} + 
    \lambda(t) \frac{\partial}{\partial x} f \\
    \frac{\mathrm{d}}{\mathrm{d} t} \frac{\partial}{\partial \dot{y}} \mathcal{L} =
    \frac{\partial}{\partial y} \mathcal{L} + 
    \lambda(t) \frac{\partial}{\partial y} f \\
    f(x,y) = \text{const}.
  \end{array}
\right.
$$

Si bien $\lambda(t)$, llamado **multiplicador de Lagrange**, es un artificio que incluimos para ayudarnos a resolver un problema matemático, resulta estár estrechamente ligado a la fuerza generalizada de vínculo.

## Una resolución con el multiplicador de Lagrange

Recordemos el Lagrangiano escrito en coordenadas cartesianas para una partícula en un espacio bidimensional

$$
\mathcal{L} (x,y,\dot{x}, \dot{y})= T - V = \frac{m}{2} (\dot{x}^2 + \dot{y}^2) - V(x,y).
$$
Resuelvo para $x$ la ecuación de Euler Lagrange modificada

$$
\begin{aligned}
    \frac{\mathrm{d}}{\mathrm{d} t} \frac{\partial}{\partial \dot{x}} \mathcal{L}
    = \frac{\partial}{\partial x} \mathcal{L} + 
    \lambda(t) \frac{\partial}{\partial x} f \\
    \frac{\mathrm{d}}{\mathrm{d} t} m \dot{x} 
    = - \frac{\partial}{\partial x} V + 
    \lambda(t) \frac{\partial}{\partial x} f \\
    m \ddot{x} 
    = - \frac{\partial}{\partial x} V + 
    \lambda(t) \frac{\partial}{\partial x} f \end{aligned}
$$

¿Qué son estos términos?
- La 2.a ley de Newton nos dice que el lado izquierdo es la suma de la componentes en $\hat{y}$ de las fuerzas actuando sobre la partícula: $\sum F_x= m \ddot{x}$.
- La fuerza causada por el potencial es $F^V_x = - \frac{\partial}{\partial x} V$.
- El tercer término tiene que tener las mismas unidades, por tanto es también una fuerza.
Sobre está partícula solo actúa la ya calculada $F_V$ y el vínculo $f$.
No hay más alternativa que admitir que este término responde a la fuerza que mantiene este vínculo

$$
    F^\text{vínculo}_x = \lambda(t) \frac{\partial}{\partial x} f.
$$

El significado del multiplicador de Lagrange $\lambda(t)$ parece ser este: multiplicado por la correspondiente $\frac{\partial}{\partial q_i} f(q_i)$ permite obtener la componente correspondiente de la fuerza de vínculo.

## Formulación E-L

### Multiplicadores de Lagrange
_Adaptado de la sección 7.5 de Stephen T. Thornton, y Jerry B. Marion. Classical Dynamics of Particles and Systems. Fifth. Thomson Brooks/Cole, 2004._


En forma general se escribe la ecuación de Euler-Lagrange con los multiplicadores del último como

$$
\boxed{
\frac{\partial}{\partial q_i} \mathcal{L} 
- \frac{\mathrm{d}}{\mathrm{d} t} \frac{\partial}{\partial \dot{q_i}} \mathcal{L} 
+ \sum_j \lambda_j(t) \frac{\partial}{\partial q_i} f_j = 0
}
\quad
\left\{
  \begin{array}{l}
  i= 1,2,\dots,n \quad \text{coordenadas generalizadas} \\
  j= 1,2,\dots,k \quad \text{condiciones cinemáticas, constreñimientos o ligaduras}
  \end{array}
\right.
$$
donde

$$
\sum_j \lambda_j(t) \frac{\partial}{\partial q_i} f_j.
$$
denota todas las fuerzas generalizadas de vínculo.

### Ecuaciones de Euler-Lagrange con fuerzas generalizadas de vínculo y no conservativas
$$
\boxed{
    \frac{\partial}{\partial q_i} \mathcal{L}
    - \frac{\mathrm{d}}{\mathrm{d}t} \frac{\partial}{\partial \dot{q}_i} \mathcal{L}
    + \sum_j \lambda_j \frac{\partial}{\partial q_i}  f_j
    + Q_i
    = 0
}.
$$
