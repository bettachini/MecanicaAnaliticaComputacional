## Momento y fuerzas generalizadas
_Adaptado de la sección 7.1 John Robert Taylor. Classical Mechanics. University Science Books, 2005._

Escribo el Lagrangiano en coordenadas cartesianas para una partícula un espacio bidimensional
$$
L (x,y,\dot{x}, \dot{y})= T - V = \frac{m}{2} (\dot{x}^2 + \dot{y}^2) - V(x,y),
$$
si lo derivara respecto a la velocidad en $\hat{x}$
$$
\frac{\partial}{\partial \dot{x}} L = \frac{\partial}{\partial \dot{x}} T = m \dot{x} = p_x,
$$
que es el **momento** en $\hat{x}$.

Si derivara respecto de la coordenada en si 
$$
\frac{\partial}{\partial x} L = - \frac{\partial}{\partial x} V = F_x,
$$
obtengo la **fuerza** causada por el potencial $V$.

Podríamos recuperar la 2.a ley de Newton
$$
\left\{
  \begin{array}{l}
    \frac{\partial}{\partial x} L =  \frac{\mathrm{d}}{\mathrm{d} t} \frac{\partial}{\partial \dot{x}} L\\
    \frac{\partial}{\partial y} L =  \frac{\mathrm{d}}{\mathrm{d} t} \frac{\partial}{\partial \dot{y}} L
  \end{array}
\right.
\iff
\left\{
  \begin{array}{l}
    F_x = m \ddot{x} = \frac{\mathrm{d}}{\mathrm{d} t} p_x\\
    F_y = m \ddot{y} = \frac{\mathrm{d}}{\mathrm{d} t} p_y
  \end{array}
\right.
$$

Si usamos coordenadas generalizadas para $L(q_i,\dot{q}_i,t)$ estas derivadas no son necesariamente momentos o fuerzas en el sentido vectorial, pero tienen un rol similar y por eso:
-  $\frac{\partial}{\partial \dot{q}_i} L$: es la componente $i$ de un **momento generalizado**
-  $\frac{\partial}{\partial q_i} L$: es la componente $i$ de una **fuerza generalizada**

### Partícula en sistema polar
Veamos en otro sistema de coordenasa que son estos momentos y fuerzas generalizadas.

En este sistema
$$
L(\rho, \dot{\rho},\varphi, \dot{\varphi}) = \frac{m}{2} \left( \dot{\rho}^2+ \rho^2 \dot{\varphi}^2 \right) - V(\rho, \varphi).
$$
Aquí el momento para la cordenada generalizada $\phi$
$$
\frac{\partial}{\partial \dot{\varphi}} L = m \rho^2 \dot{\varphi},
$$
resulta ser el **momento angular**.

Y como de Euler-lagrange tenemos que
$$
\frac{\mathrm{d}}{\mathrm{d} t} \frac{\partial}{\partial \dot{\varphi}} L = \frac{\partial}{\partial \varphi} L,
$$
este última _fuerza generalizada_ para la coordenada $\varphi$ resulta ser la variación en el tiempo del momento angular, es decir el **torque**,
$$
\tau = \frac{\mathrm{d}}{\mathrm{d} t} \left( m \rho^2 \dot{\varphi} \right) = - \frac{\partial}{\partial \varphi} V = \frac{\partial}{\partial \varphi} L.
$$

## Fuerzas de vínculo y multiplicadores de Lagrange
_Adaptado de sección 7.10
John Robert Taylor. Classical Mechanics. University Science Books, 2005._


Una ecuación que explicite un vínculo podemos escribirla como
$$
f(x,y)= \text{const.}
$$
Por ejemplo en un péndulo simple de largo $l $las coordenadas cartesianas están relacionadas por  $f(x,y)= \sqrt{x^2+ y^2}= l$.

En la
_sección $\S6$ de "Un destilado..."_
dedicada a la "Ecuación de Euler-Lagrange" se explica que:

La acción
$$
S = \delta \int_{t_1}^{t_2} L(q_i, \dot{q}_i, t) \delta q_i \mathrm{d}t, 
$$
debe ser estacionaria, es decir $\delta S = 0$,
que para este problema en (x,y) en que $x(t) \rightarrow x(t) + \delta x(t)$ e $y(t) \rightarrow y(t) + \delta x(t)$,
resulta en
$$
\delta S
=
 \int_{t_1}^{t_2}
    \left( 
    \frac{\partial}{\partial x} L \delta x
    - \frac{\partial}{\partial \dot{x}} L \delta \dot{x}
    +
    \frac{\partial}{\partial y} L \delta y
    - \frac{\partial}{\partial \dot{y}} L \delta \dot{y}
    \right)
 \mathrm{d}t
$$

Ya probamos que
$$
\delta S = \delta \int_{t_1}^{t_2} \left( 
    \frac{\partial}{\partial q_i} L
    - \frac{\mathrm{d}}{\mathrm{d} t} \frac{\partial}{\partial \dot{q}_i} L
\right) \delta q_i \mathrm{d}t, 
$$
por tanto
$$
\delta S
=
 \int_{t_1}^{t_2}
    \left( 
    \frac{\partial}{\partial x} L
    - \frac{\mathrm{d}}{\mathrm{d} t} \frac{\partial}{\partial \dot{x}} L
    \right)
    \delta x \mathrm{d}t
    + 
    \left( 
    \frac{\partial}{\partial y} L
    - \frac{\mathrm{d}}{\mathrm{d} t} \frac{\partial}{\partial \dot{y}} L
    \right)
\delta y \mathrm{d}t = 0
$$
Si esta expresión fuera cierta para variaciones arbitrarias de $\delta x$ y $\delta y$ resolveríamos ambas ecuaciones cada cual por su lado.
Pero aquí hay un vínculo entre ambas, que tambien podemos expresar en función de estas $\delta x$ y $\delta y$ como
$$
\delta f = \frac{\partial}{\partial x} f \delta x +  \frac{\partial}{\partial y} f \delta y = 0,
$$
que es nula pues estaba $f$ estaba igualada a una constante.

Así que sumar esta última al integrando de la última integral, aunque la multiplicaramos por una constante dependiente del tiempo $\lambda(t)$ no tiene porque cambiar el valor nulo de la integral
$$
\begin{aligned}
\delta S
=&
 \int_{t_1}^{t_2}
    \left( 
    \frac{\partial}{\partial x} L
    - \frac{\mathrm{d}}{\mathrm{d} t} \frac{\partial}{\partial \dot{x}} L
    + \lambda(t) \frac{\partial}{\partial x} f
    \right)
    \delta x \mathrm{d}t \\
&+ \int_{t_1}^{t_2}
    \left( 
    \frac{\partial}{\partial y} L
    - \frac{\mathrm{d}}{\mathrm{d} t} \frac{\partial}{\partial \dot{y}} L
    + \lambda(t) \frac{\partial}{\partial y} f
    \right)
\delta y \mathrm{d}t = 0.
\end{aligned}
$$

Y aquí viene el golpe de astucia, podemos variar $\lambda(t)$ hasta que la cantidad entre paréntesis que acompaña $\delta x$ se anule.
Como la suma de ambas integrales es igual a cero, esto hace asegura que lo mismo sucede con la cantidad que acompaña a $\delta y$.

Tenemos ahora solo dos ecuaciones de Euler-Lagrange  modificadas para obtener tres funciones $x(t), y(t), \lambda(t)$.
Pero como disponemos de la ecuación de la condición $f(x,y)$, tenemos tres ecuaciones y tres incógnitas:
$$
\left\{
  \begin{array}{l}
    \frac{\mathrm{d}}{\mathrm{d} t} \frac{\partial}{\partial \dot{x}} L 
    = \frac{\partial}{\partial x} L + 
    \lambda(t) \frac{\partial}{\partial x} f \\
    \frac{\mathrm{d}}{\mathrm{d} t} \frac{\partial}{\partial \dot{y}} L =
    \frac{\partial}{\partial y} L + 
    \lambda(t) \frac{\partial}{\partial y} f \\
    f(x,y) = \text{const}.
  \end{array}
\right.
$$

Si bien $\lambda(t)$, llamado **multiplicador de Lagrange**, es un artificio que incluimos para ayudarnos a resolver un problema matemático, resulta estár estrechamente ligado a la fuerza generalizada de vínculo.

### Una resolución con el multiplicador de Lagrange

Recordemos el Lagrangiano escrito en coordenadas cartesianas para una partícula libre en un espacio bidimensional
$$
L (x,y,\dot{x}, \dot{y})= T - V = \frac{m}{2} (\dot{x}^2 + \dot{y}^2) - V(x,y).
$$
Resuelvo para $x$ la ecuación de Euler Lagrange modificada
$$
\begin{aligned}
    \frac{\mathrm{d}}{\mathrm{d} t} \frac{\partial}{\partial \dot{x}} L 
    = \frac{\partial}{\partial x} L + 
    \lambda(t) \frac{\partial}{\partial x} f \\
    \frac{\mathrm{d}}{\mathrm{d} t} m \dot{x} 
    = - \frac{\partial}{\partial x} V + 
    \lambda(t) \frac{\partial}{\partial x} f \\
    m \ddot{x} 
    = - \frac{\partial}{\partial x} V + 
    \lambda(t) \frac{\partial}{\partial x} f \end{aligned}
$$

¿Qué son estos términos?
- La 2.a ley de Newton nos dice que el lado izquierdo es la suma de la componentes en $\hat{y}$ de las fuerzas actuando sobre la partícula: $\sum F_x= m \ddot{x}$.
- La fuerza causada por el potencial es $F^V_x = - \frac{\partial}{\partial x} V$.
- El tercer término tiene que tener las mismas unidades, por tanto es también una fuerza.
Sobre está partícula solo actua la ya calculada $F_V$ y el vínculo $f$.
No hay más alternativa que admitir que este término responde a la fuerza que mantiene este vínculo
$$
    F^\text{vínculo}_x = \lambda(t) \frac{\partial}{\partial x} f.
$$

El significado del multiplicador de Lagrange $\lambda(t)$ parece ser este: multiplicado por la correspondiente $\frac{\partial}{\partial q_i} f(q_i)$ permite obtener la componente correspondiente de la fuerza de vínculo.

### Formulación E-L con multiplicadores
_Adaptado de la sección 7.5 de Stephen T. Thornton, y Jerry B. Marion. Classical Dynamics of Particles and Systems. Fifth. Thomson Brooks/Cole, 2004._


En forma general se escribe la ecuación de Euler-Lagrange con los multiplicadores del último como
$$
\boxed{
\frac{\partial}{\partial q_i} L - \frac{\mathrm{d}}{\mathrm{d} t} \frac{\partial}{\partial \dot{q_i}} L - \sum_j \lambda_j(t) \frac{\partial}{\partial q_i} f_j = 0
}
\quad
\left\{
  \begin{array}{l}
  i= 1,2,\dots,n \quad \text{coordenadas generalizadas} \\
  j= 1,2,\dots,k \quad \text{condiciones cinemáticas, constreñimientos o ligaduras}
  \end{array}
\right.
$$
donde
$$
Q_i = \sum_j \lambda_j(t) \frac{\partial}{\partial q_i} f_j, 
$$
denota las fuerzas generalizadas de vínculo, pudiendo escribirse
$$
\frac{\mathrm{d}}{\mathrm{d} t} \frac{\partial}{\partial \dot{q_i}} L  - \frac{\partial}{\partial q_i} L = Q_i
$$

### Una aplicación
Si bien la ventaja de la formulación Lagrangiana era evitar tener que pensar en las fuerzas de vínculo para centrarse en obtener una dinámica, usualmente desde el punto de vista ingenieríl estas fuerzas son un interrogante de la etapa de diseño.

La figura muestra un cilíndro rodando cuesta abajo en un plano inclinado.

![](marion_fig6_7.png "Rodadura de un cilíndro bajando en un plano inclinado")

La condición de rodadura puede escribirse
$$
f(y,\theta) = y - R \theta = 0.
$$
Como rueda sobre su eje no hay que olvidar su momento de inercia $I= M R^2$ a la hora de escribir la energía cinética
$$
T = \frac{M}{2} \dot{y}^2 + \frac{I}{2} \dot{\theta}^2 = \frac{M}{2} \left( \dot{y}^2+ \frac{1}{2} R^2 \dot{\theta}^2 \right).
$$
La energía potencial solo depende del campo gravitatorio
$$
V = M g (l - y ) \sin(\alpha),
$$
donde tomé el cero de potencial en la base y $l$ es la longitud de la superficie inclinada.

Como nos interesa obtener las fuerza de vínculo no hago uso de $f(y,\theta)$ para reducir los grados de libertad a uno solo, sino que planteo
$$
\left\{
  \begin{array}{l} 
    \frac{\partial}{\partial y} L
    - \frac{\mathrm{d}}{\mathrm{d} t} \frac{\partial}{\partial \dot{y}} L
    - \lambda \frac{\partial}{\partial y} f = 0\\
    \frac{\partial}{\partial \theta} L
    - \frac{\mathrm{d}}{\mathrm{d} t} \frac{\partial}{\partial \dot{\theta}} L
    - \lambda \frac{\partial}{\partial \theta} f = 0
  \end{array}
\right. ,
$$
que resultan en
$$
\left\{
  \begin{array}{l} 
    M g \sin(\alpha)
    - M \ddot{y}
    - \lambda = 0\\
    - \frac{1}{2} M R^2 \ddot{\theta}
    + \lambda R = 0
  \end{array}
\right. .
$$
Diferenciando el vínculo dos veces $\ddot{y} = R \ddot{\theta}$ tenemos una relación útil para reemplazar en las ecuaciones que obtuvimos.
Con esto se obtiene
$$
\lambda = - \frac{1}{3} M g \sin(\alpha),
$$
y con este multiplicador de Lagrange se obtienen las fuerzas generalizadas de vínculo
$$
Q_y = \lambda \frac{\partial}{\partial y} f = \lambda = - \frac{1}{3} M g \sin(\alpha)\\
Q_\theta = \lambda \frac{\partial}{\partial \theta} f = - \lambda R = \frac{1}{3} M g R \sin(\alpha),
$$
la primera es la fuerza aplicada en el centro de masa del cilindro en la dirección $\hat{y}$, y la segunda es el correspondiente torque. 

### Sistemas no holónomos
A veces la formulación de las ecuaciones de Euler-Lagrange con multiplicadores es la única forma de usar el conocimiento que tenemos de los vínculos.

Los **vínculos holónomos** son lo que son función solo de las coordenadas generalizadas
$$
f(q_1, q_2,\dots,q_n)=0,
$$
y son los que estamos habituados a utilizar, y son sencillos de deferenciar
$$
\frac{\partial}{\partial q_1} f \mathrm{d}q_1 + \frac{\partial}{\partial q_2} f \mathrm{d}q_2 + \dots + \frac{\partial}{\partial q_n} f \mathrm{d}q_n= 0.
$$

Pero en algunas ocasiones la condición solo puede expresarse en una forma relacionada
$$
A_1(q_1,\dots,q_n) \mathrm{d}q_1 + \dots + A_n(q_1,\dots,q_n) \mathrm{d}q_n= 0,
$$
solo en diferenciales $\mathrm{d}q_n$.
Estas constituyen generalmente vínculos o condiciones **no holónomos**.

Un ejemplo de este tipo de condición es un vínculo que incluye velocidades, cuya forma más general se escribe
$$
f(q_i, \dot{q}_i, t)= 0,
$$
pues incluye las formas diferenciales $\frac{\mathrm{d}q_n}{\mathrm{d} t}$.

En contadas condiciones de integración podemos recuperar una $f(q_i,t)=0$ (excepto si el número de grados de libertad es $n=2$ en cuyo caso es siempre integrable), y verificar que en realidad se trata de un vínculo holónomo.
Un ejemplo es la forma
$$
\sum_i A_i \dot{x}_i + B = 0,
$$
si $A_i = \frac{\partial}{\partial x_i} f$ y $B = \frac{\partial}{\partial t} f$ con $f=f(x_i, t)$, que puede escribirse
$$
\sum_i \frac{\partial}{\partial x_i} f \frac{\mathrm{d}}{\mathrm{d} t} x_i + \frac{\partial}{\partial t} f = 0,
$$
pues esto es precisamente la derivada total respecto al tiempo $\frac{\mathrm{d}}{\mathrm{d} t} f$,
que puede integrarse para dar una
$$
f(x_i,t) - \text{constante} = 0,
$$
mostrando así que es holónoma.

Pero si la relación o vínculo está escrita en forma diferencial en vez de una relación algebráica podemos incorporarla directamente en la formulación con multiplicadores de Lagrange, sin tener que hacer ninguna integración previa, si tal relación se puede escribir como
$$
\sum_k \frac{\partial}{\partial q_i} f_k \mathrm{d}q_i = 0,
$$
las derivadas son la que sirven de último término acompañando a los respectivos $\lambda_k(t)$.
Esto funciona inclusive si enla condición en cuestión figura una función explicita del tiempo $\frac{\partial}{\partial t} f_k \mathrm{d} t$.
