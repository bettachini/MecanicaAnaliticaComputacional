## Fuerzas de ligadura y multiplicadores de Lagrange
>Adaptado de  
sección 7.10  
John Robert Taylor. Classical Mechanics. University Science Books, 2005.


Cuando hay restricciones a las libertades de un sistema decimos que hay una **restricción**.
Para imponer tal restricción debe haber una **ligadura** que realice la(s) **fuerza(s)** que sea(n) necesaria(s) para que se cumpla.

La restricción se puede escribir como una **función de restricción** que está restringida a un valor constante
$$
f(x,y) = \text{const.}
$$

Por ejemplo en un péndulo ideal rígido cuya cuerda tiene largo $\ell$  
![](figuras/pénduloHorizontal.png)  
las coordenadas cartesianas están restringidas a cumplir
$$
f(x,y)= \sqrt{x^2+ y^2}= \ell.
$$

### El multiplicador de Lagrange en las ecuaciones de Euler-Lagrange

En la
_sección $\S6$ de "Un destilado..."_ dedicada a la "Ecuación de Euler-Lagrange" se explica porque la dinámica del sistema es aquella para la cuál la acción

$$
S = \int_{t_1}^{t_2} \mathcal{L}(q_i, \dot{q}_i, t) \mathrm{d}t, 
$$
**debe ser estacionaria**.

Esto es equivalente a pedir que **la variación de tal acción debe ser nula**
$$
\delta S 
= \delta \int_{t_1}^{t_2} \mathcal{L}(q_i, \dot{q}_i, t) \mathrm{d}t
=  \int_{t_1}^{t_2} \delta \mathcal{L}(q_i, \dot{q}_i, t) \mathrm{d}t
= 0
$$

Variar el Lagrangiano, $\delta \mathcal{L}$ solo se logra variando las coordenadas,$\delta q_i$, y velocidades generalizadas, $\delta \dot{q}_i$.
En el problema del péndulo expresado en coordenadas $x$ e $y$ esto requeriría que
$$
x(t) \rightarrow x(t) + \delta x(t)\\
\dot{x}(t) \rightarrow \dot{x}(t) + \delta \dot{x}(t)\\
y(t) \rightarrow y(t) + \delta y(t)\\
\dot{y}(t) \rightarrow \dot{y}(t) + \delta \dot{y}(t)
$$
Contemplando estas cuatro posibles variaciones la **variación de la acción** se expresaría como

$$
\delta S
=
 \int_{t_1}^{t_2}
    \left( 
    \frac{\partial}{\partial x} \mathcal{L} \delta x
    - \frac{\partial}{\partial \dot{x}} \mathcal{L} \delta \dot{x}
    +
    \frac{\partial}{\partial y} \mathcal{L} \delta y
    - \frac{\partial}{\partial \dot{y}} \mathcal{L} \delta \dot{y}
    \right)
 \mathrm{d}t
$$
En la clase sobre la Ecuación de Euler-Lagrange ya probamos (ver sección 6 del _Apunte Lanczos_) que para cada $q_i$ esto se puede re-escribir 

$$
\delta S =  \int_{t_1}^{t_2} \left( 
    \frac{\partial}{\partial q_i} \mathcal{L}
    - \frac{\mathrm{d}}{\mathrm{d} t} \frac{\partial}{\partial \dot{q}_i} \mathcal{L}
\right) \delta q_i \mathrm{d}t, 
$$
por tanto que en este sistema descripto por las coordenadas $x$ e $y$ la **acción será estacionaria** si

$$
\delta S
=
 \int_{t_1}^{t_2}
    \left[
    \left( 
    \frac{\partial}{\partial x} \mathcal{L}
    - \frac{\mathrm{d}}{\mathrm{d} t} \frac{\partial}{\partial \dot{x}} \mathcal{L}
    \right)
    \delta x 
    + 
    \left( 
    \frac{\partial}{\partial y} \mathcal{L}
    - \frac{\mathrm{d}}{\mathrm{d} t} \frac{\partial}{\partial \dot{y}} \mathcal{L}
    \right)
\delta y 
\right]
\mathrm{d}t = 0
$$
Si esta expresión fuera cierta para variaciones arbitrarias de $\delta x$ y $\delta y$ resolveríamos ambas ecuaciones cada cual por su lado.

Pero aquí hay una restricción que relaciona ambas,
$$
f(x,y) = cte.
$$
que tambien podemos expresar en función de estas $\delta x$ y $\delta y$ como
$$
\delta f = \frac{\partial}{\partial x} f \delta x +  \frac{\partial}{\partial y} f \delta y = 0,
$$
que es nula pues si $f$ estaba igualada a una constante, su variación es nula $\delta f $.

Así que si agregáramos esta $\delta f$ al integrando de la integral para $\delta S$, aún si se la multiplicara por una constante dependiente del tiempo $\lambda(t)$, no tendría porque cambiar el valor nulo de $\delta S$

$$
\delta S
= \int_{t_1}^{t_2}
    \left[
    \left( 
    \frac{\partial}{\partial x} \mathcal{L}
    - \frac{\mathrm{d}}{\mathrm{d} t} \frac{\partial}{\partial \dot{x}} \mathcal{L}
    + \lambda(t) \frac{\partial}{\partial x} f
    \right)
    \delta x
+ 
    \left( 
    \frac{\partial}{\partial y} \mathcal{L}
    - \frac{\mathrm{d}}{\mathrm{d} t} \frac{\partial}{\partial \dot{y}} \mathcal{L}
    + \lambda(t) \frac{\partial}{\partial y} f
    \right)
\delta y 
\right]
\mathrm{d}t = 0.
$$

Y aquí viene el golpe de astucia, podemos variar $\lambda(t)$ hasta que la cantidad entre paréntesis que acompaña $\delta x$ se anule.
Como la suma de ambos terminos integrados debe ser igual a cero, hacer lo anterior asegura que lo mismo suceda con la cantidad que acompaña a $\delta y$.
Es decir que para un determinado $\lambda(t)$ se tiene

$$
\left\{
  \begin{array}{l}
    \frac{\partial}{\partial x} \mathcal{L}
    - \frac{\mathrm{d}}{\mathrm{d} t} \frac{\partial}{\partial \dot{x}} \mathcal{L}
    + \lambda(t) \frac{\partial}{\partial x} f = 0\\
    \frac{\partial}{\partial y} \mathcal{L}
    - \frac{\mathrm{d}}{\mathrm{d} t} \frac{\partial}{\partial \dot{y}} \mathcal{L}
    + \lambda(t) \frac{\partial}{\partial y} f = 0
  \end{array},
\right.
$$

Evidentemente no es posible determinar tres funciones $x(t), y(t), \lambda(t)$ con solo dos ecuaciones de Euler-Lagrange.
Se hace también necesario usar la ecuación de la condición $f(x,y)$:

$$
\boxed{
\left\{
  \begin{array}{l}
      \frac{\partial}{\partial x} \mathcal{L}
    - \frac{\mathrm{d}}{\mathrm{d} t} \frac{\partial}{\partial \dot{x}} \mathcal{L}
    + \lambda(t) \frac{\partial}{\partial x} f = 0\\
    \frac{\partial}{\partial y} \mathcal{L}
    - \frac{\mathrm{d}}{\mathrm{d} t} \frac{\partial}{\partial \dot{y}} \mathcal{L}
    + \lambda(t) \frac{\partial}{\partial y} f = 0\\
    f(x,y) - \text{const} = 0
  \end{array}
\right.
}.
$$

Si bien $\lambda(t)$, llamado **multiplicador de Lagrange**, es un artificio que incluimos para ayudarnos a resolver un problema matemático, resulta estar estrechamente ligado a la **fuerza generalizada de ligadura**.

### Un ejemplo
Para una partícula
- en un espacio bidimensional,
- sometida a un vínculo $f$,  

el Lagrangiano en cartesianas es
$$
\mathcal{L} (x,y,\dot{x}, \dot{y})= T - V = \frac{m}{2} (\dot{x}^2 + \dot{y}^2) - V(x,y),
$$
y la Euler-Lagrange para $x$ es

$$
\begin{aligned}
    \frac{\mathrm{d}}{\mathrm{d} t} \frac{\partial}{\partial \dot{x}} \mathcal{L} =
      \frac{\partial}{\partial x} \mathcal{L}
    + \lambda(t) \frac{\partial}{\partial x} f \\
    \frac{\mathrm{d}}{\mathrm{d} t} m \dot{x} =
    - \frac{\partial}{\partial x} V 
    + \lambda(t) \frac{\partial}{\partial x} f \\
    m \ddot{x} =
    - \frac{\partial}{\partial x} V
    + \lambda(t) \frac{\partial}{\partial x} f .
\end{aligned}
$$

¿Qué son estos términos?
- A la izquierda del signo igual está la $m \ddot{x}$, que según la 2.a ley de Newton es la suma de la componentes en $\hat{x}$ de las fuerzas actuando sobre la partícula, $m \ddot{x} = \sum F_x$.
- El primero a la derecha del igual $- \frac{\partial}{\partial x} V$ es la fuerza en $\hat{x}$ causada por un potencial $F^V_x = - \frac{\partial}{\partial x} V$
- El término restante debe ser otra fuerza.
Sobre ésta partícula solo actúa la ya calculada $F_V$ y la de ligadura $f$.
Por tanto, este término responde a la fuerza de ligadura que mantiene esta restricción:

$$
    F^\text{vínculo}_x = \lambda(t) \frac{\partial}{\partial x} f.
$$

Al multiplicador de Lagrange $\lambda(t)$ se lo define por lo que hace: al multiplicarlo por $\frac{\partial}{\partial q_i} f(q_i)$ se obteniene la correspondiente componente de la fuerza de ligadura.

## Forma general de Euler-Lagrange con multiplicadores de Lagrange
> Adaptado de  
sección 7.5  
Stephen T. Thornton, y Jerry B. Marion. Classical Dynamics of Particles and Systems. Fifth. Thomson Brooks/Cole, 2004.

En forma general se escribe la ecuación de Euler-Lagrange con los multiplicadores del último como

$$
\boxed{
\frac{\partial}{\partial q_i} \mathcal{L} 
- \frac{\mathrm{d}}{\mathrm{d} t} \frac{\partial}{\partial \dot{q_i}} \mathcal{L} 
+ \sum_k \lambda_k(t) \frac{\partial}{\partial q_i} f_k = 0
}
\quad
\left\{
  \begin{array}{l}
  i= 1,2,\dots,n \quad \text{coordenadas generalizadas} \\
  k= 1,2,\dots,p \quad \text{condiciones cinemáticas, de restricción o ligaduras}
  \end{array}
\right.
$$
donde
$$
\sum_k \lambda_k(t) \frac{\partial}{\partial q_i} f_k.
$$
denota todas las fuerzas generalizadas producto de las $f_k$ ligaduras.
