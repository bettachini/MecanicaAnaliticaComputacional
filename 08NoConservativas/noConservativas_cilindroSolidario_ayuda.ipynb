### Pensar en los trabajos virtuales

La fuerza del amortiguador ya nos la dieron, suponiendo un versor $\hat{e}_y$ vertical




$$
\begin{aligned}
\delta W^\text{nc}
& = \sum_j \vec{F}_j^\text{ nc} \cdot \delta \vec{r}_j = \sum_i Q_i \delta q_i \\
& = - c \dot{\vec{r}} \cdot \delta \vec{r}^\text{ nc} \\
\end{aligned}
$$

Donde $\delta \vec{r}^\text{ nc} $ será una variación virtual: de trabajos virtuales en la teórica "donde $\delta\vec{r}_j$ es el **desplazamiento virtual de la coordenada vectorial donde se aplica la fuerza externa no conservativa** $\vec{F}^{\text{ nc}}$"

La fuerza del amortiguador depende de la velocidad, esto es de la velocidad del punto rojo indicado en el diagrama de abajo.







imagen.png

La velocidad de rotación es la misma para ambos cilindros, son solidarios, entonces:

$$
\vec{F}^\text{ nc} = -c (v_x \hat{e}_x + v_y \hat{e}_y) = -c \, v_y \hat{e}_y  = -c \, (R_2 \dot{\theta}) \hat{e}_y
$$

Ahora tenemos que mirar el desplazamiento virtual:

$$
\delta \vec{r}^{\text{nc}} = \frac{\partial \, \vec{r}^{\text{ nc}}}{\partial \theta} \delta \theta = R_2 \delta \theta
$$

De acá nos queda entonces que:

$$
\begin{aligned}
\delta W^{\text{ nc}} &= \vec{F}^\text{ nc} \cdot \left( \frac{\partial \, \vec{r}^{\text{ nc}}}{\partial \theta} \right) \delta \theta = Q_{\theta} \delta \theta\\
&= -c (R_2 \dot{\theta}) R_2 \delta \theta\\
&= -c \, R_2 ^2 \, \dot{\theta} \delta \theta
\end{aligned}
$$

Esto nos dice que $ Q_{\theta} = -c \, R_2 ^2 \, \dot{\theta} $


Esta fuerza generalizada es la que tenemos que agregar en las ecuaciones de Euler-Lagrange.



$$
\frac{\partial}{\partial \theta} \mathcal{L}
- \frac{\mathrm{d}}{\mathrm{d}t} \frac{\partial}{\partial \dot{\theta}} \mathcal{L}
+ \sum_k \lambda_k \frac{\partial}{\partial \theta} f_k
+ Q_{\theta}
= 0,
$$

Acá dejé indicada la posibilidad de tener vínculos (los multiplicadores de Lagrange $\lambda_k$), si no los hubiere entonces directamente ese término no aparece. Les dejo a ustedes pensar si necesitamos o no incluirlos en este ejercicio.
