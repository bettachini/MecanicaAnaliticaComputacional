El momento angular de una masa puntual se define como
$$
\vec{L_0}(t) = \vec{r}(t) \times \vec{p}(t) = m (\vec{r} \times \vec{v} ),
$$
donde $\vec{r}$ y $\vec{p}= m \vec{v}$ son los vectores posición y momento.

Es importante acordarse que el subíndice indica que depende del punto de referencia (centro de coordenadas) que se seleccionó para definirlo.
Como está definido como un producto vectorial significa que entonces resulta ser perpendicular a los vectores que lo definen, por ejemplo si tomamos un sistema de coordenadas $\hat{e}_1,\hat{e}_2,\hat{e}_3$ con centro de coordenadas en $O$ tenemos que  
$$
\vec{r} = \rho \, \hat{e}_1 \\ 
\vec{p} =  m v \, \hat{e}_2,
$$
donde $\rho$ es la distancia en la dirección radial.
 
Entonces el momento agular resulta 
$$
\vec{L_A} = x \, \hat{e}_1  \times m v \, \hat{e}_2 = \rho m v \, (\hat{e}_1  \times \hat{e}_2) = x m v \, \hat{e_3}  
$$

Así como decíamos que si la suma de fuerzas era nula entonces el momento se conservaba, podemos encontrar un teorema equivalente para el momento angular a partir de los torques.

El torque se define como
$$
\vec{\tau}_0 = \vec{r} \times \vec{F}
$$
nuevamente, el torque se define respecto de un punto y resulta entonces perpendicular a la fuerza y al brazo de palanca $\vec{r}$ que lo define: es el vector que va desde la posición "0" hasta el punto donde se aplica la fuerza. De la segunda ley de Newton resulta entonces que
$$
\vec{\tau} = \frac{d\vec{L}}{dt}
$$
así, los torques definen los cambios de momento angular y si los torques se anulan en alguna dirección, el momento angular se conserva en ella. Se suprimió aquí la indicación del origen seleccionado.

Y de la misma forma que para las fuerzas vale:

$$
\vec{\tau}_{\text{tot}} = \sum_i \vec{\tau}_i = \frac{d\vec{L}}{dt}
$$

Hay que recordar que el brazo de fuerza llega hasta la posición donde se aplica la fuerza. Por esto era que si queremos abrir una puerta, por ejemplo, tenemos que aplicar más fuerza si la abrimos tirando cerca del eje de giro que lejos de él. Para lograr el mismo torque tenemos que aumentar la fuerza porque el brazo de fuerza cambió. 

Además, en la segunda parte de la materia vamos a mirar el momento angular y los torques aplicados a cuerpos rígidos. De las materias básicas de física sabemos que vale:
$$
\vec{L} = I \vec{\omega}
$$
que utilizamos ya en algunos ejercicios.
