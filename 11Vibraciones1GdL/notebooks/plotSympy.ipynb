## Forzado periódico

```python
import sympy as sym
import matplotlib.pyplot as plt
plt.rcParams['figure.figsize'] = 10, 5 # tamaño figuras sympy.plot

x = sym.symbols('x')
a = sym.plot(sym.cos(x), show=False)
b = sym.plot(sym.sin(x), show=False)
a.append(b[0])
```
