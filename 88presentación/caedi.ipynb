# Herramientas, competencias y habilidades

- Pizarrón + papel (s. XIX) $\Rightarrow$ Ejecutar código en línea (s. XXI)
    - **Sólo** resoluciones analíticas $\Rightarrow$ Explotar el **<font color='red'>cálculo numérico</font>**
    - Análisis **cualitativo** $\Rightarrow$ **Visualización cuantitativa** de la dinámica
    - Sistemas de complejidad **limitada** $\Rightarrow$ Dispositivos  **realistas**

- Basado en recursos **en línea** $\Rightarrow$ Cursos remotos o de **<font color='red'>aula híbrida</font>**
    - contratadas por la UNLaM: MS Teams (hay alternativas de software libre)
    - de acceso público y gratuito: GitHub, Google Colaboratory

- Con el trabajo con un **<font color='red'>lenguaje de programación</font>** los alumnos ganan experiencia en
    - competencias digitales $\Rightarrow$ alta demanda del sector productivo
    - trabajo remoto colaborativo $\Rightarrow$ deslocalización de la mano de obra
    - uso de software libre $\Rightarrow$ facilita el autoempleo con bajo capital
