## Why is this course called _Computational analytical mechanics_?

- The official name is _Analytical mechanics_
    - La física de un sistema de cuerpos puede resolverse en forma **vectorial**, como hicieron en Física I,
    - o haciendo balance de **energías** (cinética, potencial, etc), como también hicieron en FI.
    - Este último enfoque, el **analítico** es mucho **más poderoso** de lo que vieron en FI.
    - Mejor que explicar ahora que esto **ahorra mucho trabajo** es aprenderlo en las próximas clases y **disfrutar** de esforzarse menos.

## Cálculo numérico

/home/vbettachini/documents/universitet/unlam/MecanicaAnaliticaComputacional/01Vectorial/estaFigs/vbush_overvw.gif

[El 'analizador diferencial' inventado por el ingeniero Vannevar Bush y finalizado en 1931 por su equipo en el Massachusetts Institute of Technology (MIT) en Boston, Estados Unidos de América.](https://ub.fnwi.uva.nl/computermuseum/vbush_tbl.html)

- El análisis numérico, que conocen de la asignatura _Cálculo numérico_, existe desde [tiempos babilónicos](https://en.wikipedia.org/wiki/Numerical_analysis) cuando lo realizaban en tabletas de arcilla.
    - Hacer cálculo numérico **en papel**, es seguir en los tiempos de la arcilla, **nadie en su sano juicio** lo hace manualmente.
- Antes de la computadora digital se inventaron máquinas mecánicas y/o eléctricas para hacer automatizar el cálculo numérico
    - No automatizarle lo hace laborioso y propenso a errores.
    - El "analizador diferencial" mostrado arriba es un ejemplo que recibe su nombre por resolver problemas del **análisis diferencial** (derivadas, integrales), el empleo más usual del cálculo numérico.
- Desde **mitad del s.XX** el cálculo numérico se realiza en **computadoras** digitales.
    - La herramienta es fundamental en la ingeniería.
    - No enseñar a los alumnos como explotar la computadora para esto es atentar contra su formación.

## Álgebra computacional

/home/vbettachini/documents/universitet/unlam/MecanicaAnaliticaComputacional/01Vectorial/estaFigs/ucarecdn.jpeg

Integración simbólica en [Macsyma](https://en.wikipedia.org/wiki/Macsyma), creado en 1968 es uno de los sistemas de álgebra computacional más antigüos aún en uso.

- A principios de los 1960s varios fisicos, como el Nobel [Martinus Veltman](https://en.wikipedia.org/wiki/Schoonschip), escribieron en FORTRAN o LISP [sistemas de álgebra computacional](https://en.wikipedia.org/wiki/Computer_algebra_system) para:
    - manipular expresiones matemáticas de igual forma que se hace manualmente
    - realizar automáticamente simplificaciones
    - hacer transformadas de Fourier y Laplace
    - calcular expansiones en series, y la suma de sus truncamientos
    - hacer integración indefinida y con límites,
    - realizar operaciones matriciales, etc.

## Mecánica analítica **computacional**

- Quienes cursan _Mecánica analítica computacional_...
    - ...ya aprobaron la __escuela primaria__:
        - Demostraron que entienden y pueden hacer cálculos arítméticos manualmente.
        - Hoy resuelven sus cálculos aritméticos más rápido y con menos errores en una __calculadora de bolsillo__.
    - ...ya aprobaron en la __universidad__ asignaturas de álgebra y análisis matemático:
        - Demostraron que entienden y pueden hacer cálculos integral y diferencial manualmente así como cálculos espaciales (vectoriales) y matriciales.
        - A partir de hoy realizarán los mísmos más rápido y con menos errores utilizando un **sistema de álgebra computacional**.
- La computadora puede hacer más que cálculo numérico para aliviar el esfuerzo que el ingeniero al resolver un problema matemático.
    - Puede también realizar operaciones de álgebra y análisis simbólicos
    - En este curso la utilizaremos también para obtener las ecuaciones diferenciales para la dinámica de sistemas de cuerpos rígidos.

## Código para resolver problemas de mecánica

- Escribiremos código del lenguaje de programación general [Python](https://www.python.org/)
- Escribir código $\neq$ programar
    - **Programar**: Saber cómo analizar y modelar matemáticamente un problema (lógica), idear y comparar alternativas y saber calcular cual es la mejor considerando contexto de uso, restricciones, arquitectura del sistema informático, así como cuestiones de interfaz con el operador humano, etc.
        - Requiere **bastante** estudio y esfuerzo.
        - Todo estudiante de ingeniería **puede** aprender a programar.
    - **Escribir código**: Plasmar una solución en un lenguaje de programación particular.
        - Requiere **poco** estudio y esfuerzo.
        - Todo estudiante de ingeniería **DEBE** aprender a escribir código.
- Usaremos un mínimo de funciones de bibliotecas
    - [SymPy](https://www.sympy.org/es/index.html) para álgebra y análisis simbólico
    - [NumPy](https://numpy.org/) y [SciPy](https://scipy.org/) para cálculo numérico
    - [matplotlib](https://matplotlib.org/) para generar representaciones visuales de los resultados

## [Aula invertida](https://es.wikipedia.org/wiki/Aula_invertida): cambia el uso del tiempo

../01Vectorial/figuras/diagramaTiempo2.png

- Teoría
    - se lee, e intenta aplicar, **antes de la clase**
    - las dudas se aclaran **en clase**
    - si surgen nuevas dudas se pueden hacer consultas adicionales **tras la clase**
- Ejercicios
    - se inician **antes de la clase**
    - se terminan **en clase**
    - correcciones sugeridas por el docente **tras la clase**

| Tiempo    | Modalidad tradicional | Aula invertida |
| -------- | ------- |------- |
| Sincrónico con los docentes<br />(en clase)  | Énfasis en la exposición de nueva teoría.<br />Solo un resto para atender consultas. |  Énfasis en terminar ejercicios de la semana.<br /> Cuando hay dificultad con uno el docente está a disposición. |
| Asincrónico<br />(en casa o donde se pueda) | Una parte es repaso de teoría.<br />La mayor parte es realizar los ejercicios.<br />Cuando hay dificultad con uno no hay docente a quién recurrir. | Buena parte es asimilar nueva teoría.<br /> Iniciar la resolución de ejercicios.

## Organización del aula invertida

- Esta modalidad requiere **mucha regularidad** por parte del alumno
- Si se espera al encuentro sincrónico (clase) para
  - leer teoría $\implies$ **no alcanza el tiempo**
  - empezar los ejercicios $\implies$ **no alcanza el tiempo**
  - ponerse _al día_ con lo adeudado de la semana panterior $\implies$ **no alcanza el tiempo**
- Para ayudar a manterse al día habrá una **evaluación semanal**
  - **Nueva guía** de ejercicios cada semana
  - **Entrega obligatoria** de dos o más ejercicios vía Teams a las **21:50 hs** en cada encuentro sincrónico
- Por supuesto que puede fallarse en uno o más de los ejercicios que se entregan.
  - es posible que haya faltado comprensión sobre algún aspecto de teoría $\implies$ consulta y/o repaso personal de teoría
  - pero tal falla debe corregirse en una **segunda entrega**.

### Asistencia
- Se toma lista.

### Entregas de ejercicios
- Cada unidad de la materia tiene una guía de ejercicios, de los cuales varios serán solicitados para la entrega, éstos les aparecerán en _Teams_ bajo _Tareas_.
- Para aprobar la materia DEBEN realizar las entregas semanales de ejercicios.

- Las clases son para consultar y avanzar en los ejercicios, no duden en preguntarnos todo lo que requieran, de forma tal de que puedan aprovecharnos al máximo para evacuar sus dudas.
    - Es muy importante que aprovechen las clases para las consultas y que también nos escriban por Teams durante la semana consultando lo que no les salga.
- El día que se presenta la unidad podrán comenzar a mirarlos y resolverlos, tendrán la siguiente clase para realizar consultas adicionales y deberán ser entregados indefectiblemente antes de que comience la siguiente.

### Esquema de entregas
> La guía de la clase N se publica en la clase N-1 $\rightarrow$ en clase N pueden consultar $\rightarrow$ al finalizar la clase N todos los ejercicios solicitados deben ser entregados.

- Pasado el día de entrega $\implies$ penalización sobre la nota
    - Por ello es importante que intenten avanzar en ellos, no es necesario que estén perfectos, una vez los envíen se los corregiremos y les pediremos las modificaciones necesarias para que puedan avanzar hasta que lo terminen satisfactoriamente.
    - Cuando termine la cursada entonces deberían haber logrado hacer **TODOS** los ejercicios que les enviamos como _Tarea_.
