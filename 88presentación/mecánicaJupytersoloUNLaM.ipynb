## Mecánica General: la computadora eje de la enseñanza en línea

- Clase de teoría:
    - **Pizarrón** $\implies$ Presentaciones con **código** (**reutilizable por los alumnos**)
    - Alumnos concentrados en transcribir un pizarrón $\Rightarrow$ concentrados en el tema de la clase. 
    - Cada palabra del docente revisable a toda hora: **video** de todas las clases
    - Y siempre se puede garabatear en un _blackboard_ virtual para aclarar algo o corregir una ecuación.

- Ejercicios:
    - **Cuaderno** $\implies$ Con **código** se resuelven los **cálculos**:
        - **simbólicos**: totalmente automático, incluyendo resolución de ecuaciones diferenciales
        - **numéricos**: simulación de los sistemas basados en la integración numérica
    - Gerenciarlos a través de los **_learning management systems_** (e.g. MIeL, Moodle, Google Classrooms, MS Teams)
        - Asignar tareas individualmente, hacer seguimiento y proveer corrección en línea
        - Proveer ejercicios de ejemplo, con **código** (**reutilizable**)

# ¿Qué puede incluirse en un cuaderno de Jupyter?

- Texto y ecuaciones con símbolos $\LaTeX$
- Código ejecutable en muchos lenguajes (Python, C, R, etc.)
    - Cálculo simbólico y numérico
    - Gráficos estáticos y animaciones de los resultados del cálculo
    - Adquisición de datos y procesamiento (**¡atención laboratorio!**)
    - Modelos de simulación interactivos

# En resumen este curso de Mecánica General

- Temática: No para estudiantes de física $\Rightarrow$ específica para estudiantes de ingeniería mecánica
    - No se limitaron a sistemas triviales $\Rightarrow$ **atacaron** sistemas mecánicos
- Paradigma: No pizarrón + cuaderno (s. XIX) $\Rightarrow$ manejar código en línea (s. XXI)
