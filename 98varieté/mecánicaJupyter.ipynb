# La computadora como herramienta central de enseñanza

- Teoría: **adios al pizarrón** $\implies$ todo en presentaciones pero con código (**remixeables**)
    - Grabaciones de video de todas las clases: disponibles como referencia a toda hora
        - **Basta** de alumnos concentrados en transcribir lo que se escribe en un pizarrón. 
    - ¿Alguna consulta? Existe el _blackboard_ virtual para aclarar algo o corregir una ecuación.

- Ejercicios: **adiós al cuaderno** $\implies$ todos los cálculos en código:
    - simbólico: totalmente automático, incluyendo resolución de ecuaciones diferenciales
    - numérico: simulación de los sistemas basados en la integración numérica
    - Gerenciarlos a través de los **learning management systems** (i.e. MIeL, Moodle, Google Classrooms, MS Teams)
        - Asignar tareas individualmente, hacer seguimiento y proveer corrección en línea
        - Proveer ejercicios de ejemplo, con **código** (**remixeables**)

# ¿Qué puede incluirse en un cuaderno de Jupyter?

- Imágenes, video, sonido y enlaces web
- Texto y ecuaciones con símbolos $\LaTeX$
- Código ejecutable en muchos lenguajes (Python, C, R, etc.)
    - Cálculo simbólico y numérico
    - Gráficos estáticos y animaciones de los resultados del cálculo
    - Adquisición de datos y procesamiento (**¡atención laboratorio!**)
    - Modelos interactivos

# En conclusión

- Cursada en línea de Mecánica General que no es una mala adaptación de un curso en pizarrón
- Temática: para estudiantes de física $\Rightarrow$ específica para estudiantes de ingeniería mecánica
- Paradigma:  pizarrón + cuaderno (s. XIX) $\Rightarrow$ manejar código en línea (s. XXI)
- Alcance: limitado a resoluciones analíticas $\Rightarrow$ explotar el cálculo numérico (¡Para algo cursaron esa materia!)

- La cursada a distancia con esta modalidad **ES MEJOR** que la que podríamos brindar en el aula
