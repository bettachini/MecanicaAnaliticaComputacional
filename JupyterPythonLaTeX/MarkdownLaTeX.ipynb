# Notación Markdown

Los documentos de Jupyter se llaman cuadernos cuadernos (notebooks) se dividen internamente en **celdas**.
Se escribe en las celdas de
- **Código** sentencias en un lenguage de programación, e.g.Python
- **Texto** o **Markdown** no solo notas o explicaciones en texto plano, sino también se puden incluir, como en cualquier procesador de texto, enlaces a archivos en internet o en la misma computadora, fórmulas matemáticas, tablas o imágenes.

__Markdown__, que significa "dejar por escrito", es también un código.
Si hacen doble click sobre este texto podrán verlo y notarán como se indican encabezados o énfasis (*itálica*, **negrita**):
- __negrita__ se logra escribiendo 
` __negrita__ ` o `**negrita**`
- _italica_  se logra escribiendo 
` _italica_ ` o `*italica*`
- encabezados con signos de numeral `#` (hashtags) , dependiendo de su cantidad el nivel obtenido.
```markdown
# Encabezado de primer nivel, H1

## Encabezado de segundo nivel, H2

###### Encabezado de sexto nivel, H6
```

Mucho de lo que se publica en internet se escribe en Markdown, e.g. los artículos de Wikipedia.
El Portal Nacional de Datos Abiertos publicó una muy concisa [guía extendida en castellano de Markdown](https://datosgobar.github.io/portal-andino/markdown-guide/).

# Notación $\LaTeX$

## Una "nota" sobre la "notación" (¡cuac!).

Los cuadernos (notebook) **Jupyter** permiten utilizar la escritura de expresiones matemáticas en notación [$\LaTeX$](https://www.latex-project.org), un sistema de preparación de documentos en el que todo, inclusive las ecuaciones, se escribe como texto plano.

Por ejemplo el símbolo que designa la velocidad angular, la **phi** del alfabeto griego, se puede lograr escribiendo `\varphi` logrando que se muestre $\varphi$, o con su otra grafía `\phi` logrando $\phi$.

$\LaTeX$ permite escribir más cosas en formato matemático que símbolos sueltos.
Expresiones matemáticas complejas, e.g. $\sqrt{ \frac{\sin(\alpha^2 ) }{2} }$ pueden presentarse en la línea de texo englobando lo escrito entre símbolos de moneda como `$ código de la ecuación $ `.

Al ejecutar la celda `markdown` se reemplaza lo escrito en notación $\LaTeX$ por el equivalente en la notación matemática usual.

Si la expresión es larga, o central en lo que se explica, puede escribirse en una línea aparte englobandola en un par de `$$` y aparecerá centrada:
```
$$
    \sqrt{ \frac{\sin(\alpha^2 ) }{2} }.
$$
```
da por resultado
$$
    \sqrt{ \frac{\sin(\alpha^2 ) }{2} }.
$$

Admeás también podemos sumarle numeración utilizando `\tag{#}`, por ejemplo
```
$$
    \sqrt{ \frac{\sin(\alpha^2 ) }{2} }   \tag{1}
$$
```
resulta en 
$$
    \sqrt{ \frac{\sin(\alpha^2 ) }{2} }   \tag{1}
$$
